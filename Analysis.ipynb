Ekstraksi dilakukan secara terstruktur dengan:

Pendekatan ini memastikan bahwa data yang dianalisis benar-benar mencerminkan opini pengguna secara organik dan tidak bias terhadap sampel tertentu.

````python
# Libraries
import os
from pathlib import Path

import pandas as pd
from dotenv import load_dotenv
from googleapiclient.discovery import build
````

````python
# 1.0) Konfigurasi dan Setup
load_dotenv()

YOUTUBE_API_KEY = os.getenv("YOUTUBE_API_KEY")
if not YOUTUBE_API_KEY:
    raise EnvironmentError("YOUTUBE_API_KEY belum diset di file .env")

PROJECT_ROOT = Path(".").resolve()
OUTPUT_DIR = PROJECT_ROOT / "dungeon" / "raw_dataset"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

VIDEO_LIST = [
    {
        "id": "OHGnsJYi4tQ",
        "channel": "GadgetIn",
        "title": "Saya suka sama harga Redmi Note 15 Pro 5G!",
    },
    {
        "id": "te4V1vw2pRY",
        "channel": "Jagat Review",
        "title": "Smartphone 3 Jutaan Desain Keren, Body Tipis, dan All Rounder dari Redmi! Review Redmi Note 15 5G",
    },
    {
        "id": "r1aEXJYypx8",
        "channel": "Youtuber Cupu",
        "title": "Redmi Note 15 Pro 5G: Spek Naik, Tapi Harga?",
    },
]
````

````python
# 1.1) Proses ekstraksi komentar dari YouTube
print("Memulai Ekstraksi Data Multi Channel\n")

def extract_youtube_comments(video_id, api_key, max_comments=7000):
    youtube = build("youtube", "v3", developerKey=api_key)

    comments = []
    next_page_token = None
    fetched = 0

    try:
        while fetched < max_comments:

            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                maxResults=100,
                pageToken=next_page_token,
                textFormat="plainText",
                order="time"
            )

            response = request.execute()

            for item in response.get("items", []):

                snippet = item["snippet"]["topLevelComment"]["snippet"]

                comments.append({
                    "video_id": video_id,
                    "author": snippet.get("authorDisplayName"),
                    "text": snippet.get("textDisplay", "").strip(),
                    "like_count": snippet.get("likeCount", 0),
                    "published_at": snippet.get("publishedAt")
                })

                fetched += 1

                if fetched >= max_comments:
                    break

            next_page_token = response.get("nextPageToken")

            if not next_page_token:
                break

    except Exception as e:
        print(f"Error pada video {video_id}: {e}")

    return pd.DataFrame(comments)

all_dfs = []

for video in VIDEO_LIST:

    print(f"-> Processing: {video['channel']}")

    df_vid = extract_youtube_comments(
        video["id"],
        YOUTUBE_API_KEY,
        max_comments=7000
    )

    if not df_vid.empty:

        df_vid["source_channel"] = video["channel"]
        df_vid["video_title"] = video["title"]

        all_dfs.append(df_vid)

        print(f"(Berhasil scrap {len(df_vid):,} komentar)\n")

        # simpan per video
        safe_name = video["channel"].replace(" ", "_")
        path = OUTPUT_DIR / f"raw_{safe_name}.csv"

        df_vid.to_csv(path, index=False, encoding="utf-8-sig")

    else:
        print("(Gagal / tidak ada komentar)\n")
````

````python
# 1.2) Menggabungkan semua data dan simpan ke file
if all_dfs:

    df_raw = pd.concat(all_dfs, ignore_index=True)

    print(f"TOTAL DATA TERKUMPUL (semua channel): {len(df_raw):,} komentar")

    combined_path = OUTPUT_DIR / "combined_raw_data.csv"

    df_raw.to_csv(combined_path, index=False, encoding="utf-8-sig")

else:
    print("Gagal mengambil data.")
````

### **2. Text Preprocessing**

Tahap text preprocessing dilakukan untuk mengubah komentar mentah (raw text) menjadi format yang lebih bersih, konsisten, dan siap dianalisis oleh model NLP.

Komentar YouTube umumnya mengandung bahasa informal, emoji, singkatan, serta karakter yang tidak terstruktur. Tanpa pembersihan, hal ini dapat menurunkan akurasi model dan menyebabkan interpretasi yang tidak konsisten.

Proses preprocessing dilakukan secara sistematis melalui beberapa langkah berikut:
1. `Slang Dictionary`, Mengubah bahasa informal menjadi bentuk baku (misal: “gk” → “tidak”, “renot” → “redmi note”). Langkah ini menjaga konsistensi makna dan membantu model memahami konteks secara lebih akurat.
2. `Demoji`, Mendeteksi dan mengekstrak emoji agar dapat diproses secara terstruktur. Emoji tidak langsung dihapus, tetapi dinormalisasi agar tidak dianggap sebagai token acak.
3. `Menghapus URL & Mention`, Menghilangkan tautan dan mention yang tidak berkontribusi terhadap sentimen atau opini produk.
4. `Normalisasi Huruf Berulang`, Mengubah kata seperti “baguuuus” menjadi “bagus” agar tidak dianggap sebagai token berbeda oleh model.
5. `Replace Phrase Slang`, Menstandarkan frasa slang multi-kata agar tidak terpecah saat proses tokenisasi.
6. `Menghapus Simbol Non-Relevan`, Menghilangkan karakter non-alfabet yang bersifat noise dan tidak memiliki makna semantik.
7. `Token Slang Handling`, Memastikan kata slang dikenali sebagai token yang benar sebelum masuk ke tahap modeling.
8. `Hapus Spasi Ganda`, Membersihkan teks agar konsisten dan menghindari error saat tokenisasi.
9. `Filter Komentar Pendek`,Menghapus komentar sangat pendek (kurang dari 4 kata) karena umumnya tidak memberikan opini bermakna untuk analisis sentimen.

Text preprocessing adalah tahapan yang sangat penting karena dapat meningkatkan:
- Konsistensi bahasa
- Kualitas representasi teks
- Stabilitas tokenisasi
- Akurasi model sentimen

>Kualitas output model sangat bergantung pada kualitas input teks. Oleh karena itu, preprocessing menjadi fondasi penting sebelum masuk ke tahap analisis sentimen dan fine-tuning.

````python
# 2.2) Regex Patterns untuk Cleaning
RE_URL_MENTION = re.compile(r"http\S+|www\S+|@\w+")
RE_REPEAT = re.compile(r"(.)\1{2,}")  # >= 3 repetisi
RE_NON_TEXT = re.compile(r"[^a-z0-9\s]")  # angka tetap dipertahankan
RE_MULTI_SPACE = re.compile(r"\s+")

PHRASE_SLANG = {k: v for k, v in slang_dict.items() if " " in k}
WORD_SLANG = {k: v for k, v in slang_dict.items() if " " not in k}

phrase_patterns = sorted(PHRASE_SLANG.keys(), key=len, reverse=True)
RE_PHRASES = [
    (re.compile(rf"\b{re.escape(p)}\b"), PHRASE_SLANG[p])
    for p in phrase_patterns
]
````

````python
# 2.3) Cleaning Function
def clean_text_advanced(text, keep_numeric=True, max_repeat=2):
    if not isinstance(text, str) or not text.strip():
        return ""

    text = text.lower().strip()

    # Demojize / remove emoji
    text = demoji.replace(text, " ")

    # Hapus URL & mention
    text = RE_URL_MENTION.sub(" ", text)

    # Normalisasi huruf berulang
    def _reduce(m):
        ch = m.group(1)
        return ch * max_repeat

    text = RE_REPEAT.sub(_reduce, text)

    # Replace slang berbentuk phrase dulu
    for pattern, repl in RE_PHRASES:
        text = pattern.sub(repl, text)

    # Hapus simbol
    if keep_numeric:
        text = RE_NON_TEXT.sub(" ", text)
    else:
        text = re.sub(r"[^a-z\s]", " ", text)

    # Replace slang single-word
    words = text.split()
    fixed_words = [WORD_SLANG.get(w, w) for w in words]
    text = " ".join(fixed_words)

    # Rapikan spasi
    return RE_MULTI_SPACE.sub(" ", text).strip()
````

````python
# 2.4) Eksekusi Cleaning
df = df_raw.copy()
df["clean_text"] = df["text"].apply(clean_text_advanced)

print("Cleaning text selesai.")
````

````python
# 2.5) Filter komentar pendek (< 4 kata)
MIN_WORDS = 4
before_filter = len(df)

df = df[df["clean_text"].str.split().str.len() >= MIN_WORDS].reset_index(drop=True)

after_filter = len(df)
dropped = before_filter - after_filter

print(f"Filter komentar pendek (< {MIN_WORDS} kata) selesai.")
print(f"Sebelum filter : {before_filter:,}")
print(f"Sesudah filter : {after_filter:,}")
print(f"Terhapus       : {dropped:,}\n")
````

````python
# 3.1.3) Compile Patterns
NEG_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["negative"]]
POS_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["positive"]]

def _has_match(patterns, text):
    return any(p.search(text) for p in patterns)

def _normalize_label(label: str) -> str:
    label = str(label).lower().strip()

    if label in {"positive", "neutral", "negative"}:
        return label

    mapping = {
        "label_0": "negative",
        "label_1": "neutral",
        "label_2": "positive"
    }
    return mapping.get(label, "neutral")
````

````python
# 3.1.4) Hybrid Sentiment Function
def analyze_hybrid_sentiment(text: str, override_threshold: float = 0.90):
    if not isinstance(text, str) or not text.strip():
        return "neutral", 0.0

    t = text.lower().strip()

    # A. Prediksi model
    try:
        res = sentiment_pipeline(t, truncation=True, max_length=512)[0]
        label = _normalize_label(res.get("label", "neutral"))
        score = float(res.get("score", 0.0))
    except Exception:
        return "neutral", 0.0

    # B. Keyword boosting
    has_neg = _has_match(NEG_PATTERNS, t)
    has_pos = _has_match(POS_PATTERNS, t)

    # Negatif lebih tegas untuk override
    if has_neg:
        if (label != "negative") or (score < override_threshold):
            return "negative", 0.95

    # Positif override jika tidak ada sinyal negatif
    if has_pos and not has_neg:
        if (label != "positive") or (score < override_threshold):
            return "positive", 0.95

    return label, score
````

````python
# 3.1.5) Run Analysis
print("Menganalisis sentimen...")
results = df["clean_text"].apply(analyze_hybrid_sentiment)

df["sentiment_label"] = results.apply(lambda x: x[0])
df["sentiment_score"] = results.apply(lambda x: x[1])

print("Analisis sentimen selesai.\n")
````

````python
# 3.2.2.1) HELPERS & PROMPT
def extract_json_object(text: str):
    if not isinstance(text, str): return None
    text = text.strip()
    text = re.sub(r"^```json\s*", "", text)
    text = re.sub(r"^```\s*", "", text)
    text = re.sub(r"\s*```$", "", text)
    try:
        return json.loads(text)
    except:
        match = re.search(r"\{.*\}", text, flags=re.DOTALL)
        if match:
            try: return json.loads(match.group(0))
            except: return None
    return None

def build_prompt(text: str):
    return f"""
Anda adalah pakar anotasi data untuk proyek analisis sentimen ulasan gadget di Indonesia.

========================
OBJEK ANALISIS UTAMA
========================
Produk Target: Xiaomi Note 15 Pro 5G rilis 2026.
Sentimen harus dinilai berdasarkan bagaimana komentar tersebut berdampak pada citra atau daya tarik Xiaomi Note 15 Pro 5G.

========================
ATURAN GOLD LABELING (LOGIKA KHUSUS)
========================

1. ATURAN "MENDING" (NEGATIF): 
   Setiap komentar yang menyebut "Mending [Produk/Brand Lain]" dikategorikan NEGATIVE. 
   Ini termasuk brand pesaing (Samsung, Tecno, Infinix) maupun seri internal Xiaomi lainnya (Xiaomi T Series, Poco) yang dianggap lebih layak beli daripada Note 15.
   Contoh: "Mending Xiaomi 14T sih" -> Label: negative.

2. ATURAN PENYESALAN WAKTU / BUYER'S REMORSE (NETRAL): 
   Komentar yang mengekspresikan penyesalan karena "baru saja membeli seri lama" (seperti Note 13 atau 14) sebelum Note 15 rilis, dikategorikan NEUTRAL. 
   Kecuali jika mereka secara eksplisit menghujat Note 15.
   Contoh: "Sue bener baru beli Note 14 malah keluar yang 15" -> Label: neutral.

3. ATURAN TESTIMONI SERI LAMA (NETRAL): 
   Cerita panjang lebar tentang Note 13/12 yang masih awet tanpa memuji Note 15 adalah NEUTRAL. User hanya berbagi pengalaman pribadi, bukan menilai produk target.

========================
DEFINISI LABEL
========================

- POSITIVE: 
  Memuji fitur Note 15, menunjukkan niat beli, atau menganggap Note 15 lebih baik dari pesaingnya.

- NEUTRAL: 
  Pertanyaan spesifikasi, sapaan ("Hadir bang"), diskusi teknis tanpa opini, atau masalah timing pembelian (poin 2 & 3 di atas).

- NEGATIVE: 
  Kritik fisik/software Note 15, mengeluh harga mahal, atau mempromosikan produk lain (poin 1 di atas).

========================
KATEGORI NOTES (PILIH SALAH SATU)
========================
- Desain/Hardware (Modul kamera, build quality)
- Kamera (Kualitas foto/video)
- Baterai & Charging
- Performa (Chipset, gaming)
- Software/UI (Iklan, bug, update)
- Harga/Value (Worth it atau tidak)
- Perbandingan Kompetitor (Mending brand/seri lain)
- Testimoni Seri Lama (Pengalaman pakai HP lama)
- Buyer's Remorse (Menyesal baru beli HP lain)
- Pertanyaan/Konten (Tanya harga, request review)
- Opini Umum/Lainnya

========================
FORMAT OUTPUT (JSON SAJA)
========================
{{
  "assist_gold_label": "positive|neutral|negative",
  "assist_gold_notes": "Kategori: Deskripsi singkat alasan"
}}

========================
KOMENTAR USER
========================
\"\"\"{text}\"\"\"
""".strip()
````

````python
# 3.2.2.2) Setiap komentar akan diproses dengan fungsi pembungkus untuk memungkinkan eksekusi paralel menggunakan ThreadPoolExecutor. 
# Fungsi ini akan menangani pemanggilan API, parsing respons, dan penanganan error secara ringkas.
def llm_assist_label(index, text):
    """Fungsi pembungkus untuk diproses secara paralel"""
    if not isinstance(text, str) or not text.strip():
        return index, "neutral", "Kosong/Tidak Relevan"

    try:
        response = client.models.generate_content(
            model=MODEL_NAME,
            contents=build_prompt(text)
        )
        raw_text = getattr(response, "text", "") or ""
        parsed = extract_json_object(raw_text)

        if not parsed:
            return index, "neutral", "Fallback parsing gagal"

        label = str(parsed.get("assist_gold_label", "neutral")).strip().lower()
        notes = str(parsed.get("assist_gold_notes", "Opini Umum")).strip()
        
        if label not in VALID_LABELS: label = "neutral"
        return index, label, notes

    except Exception as e:
        # Menangani limit atau koneksi secara singkat
        return index, "neutral", f"Error: {str(e)[:50]}"
````

````python
# 3.2.2.3) Eksekusi Paralel dengan ThreadPoolExecutor
TOTAL_ROWS = len(df_label)
start_time = time.time()
start_timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

print("======================================")
print("LLM PARALLEL LABELING STARTED")
print(f"Model      : {MODEL_NAME}")
print(f"Workers    : {MAX_WORKERS} threads")
print(f"Total Data : {TOTAL_ROWS:,} komentar")
print(f"Start Time : {start_timestamp}")
print("======================================\n")

# Penampung hasil sementara (menggunakan list dengan ukuran tetap agar urutan terjaga)
results_labels = [None] * TOTAL_ROWS
results_notes = [None] * TOTAL_ROWS

with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    # Submit semua baris ke thread pool
    futures = {
        executor.submit(llm_assist_label, i, row['text']): i 
        for i, row in df_label.iterrows()
    }
    
    completed_count = 0
    for future in as_completed(futures):
        idx, label, notes = future.result()
        results_labels[idx] = label
        results_notes[idx] = notes
        
        completed_count += 1
        
        # Notifikasi biasa setiap 25 data
        if completed_count % 25 == 0 or completed_count == TOTAL_ROWS:
            pct = (completed_count / TOTAL_ROWS) * 100
            print(f"[{datetime.now().strftime('%H:%M:%S')}] Progres: {completed_count}/{TOTAL_ROWS} ({pct:.1f}%)")

# Masukkan ke DataFrame
df_label["assist_gold_label"] = results_labels
df_label["assist_gold_notes"] = results_notes

# Mengisi gold_label dan gold_notes dengan hasil assist jika gold_label masih kosong atau tidak valid
for col_lbl, col_ast in [("gold_label", "assist_gold_label"), ("gold_notes", "assist_gold_notes")]:
    if col_lbl in df_label.columns:
        df_label[col_lbl] = df_label[col_lbl].fillna("")
        mask = df_label[col_lbl].astype(str).str.strip() == ""
        df_label.loc[mask, col_lbl] = df_label.loc[mask, col_ast]

# Save
df_label.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")

end_time = time.time()
elapsed_min = (end_time - start_time) / 60

print("\n======================================")
print("LLM LABELING COMPLETED")
print(f"Duration     : {elapsed_min:.2f} menit")
print(f"Output File  : {OUTPUT_PATH}")
print("======================================")

print("\nDistribusi Hasil:")
print(df_label["assist_gold_label"].value_counts())
````

Model LLM digunakan untuk memberikan rekomendasi label sentimen awal serta interpretasi konteks komentar sebelum dilakukan verifikasi oleh annotator manusia.

Untuk meningkatkan efisiensi proses labeling dan meminimalkan dampak network latency dari pemanggilan API, proses anotasi dijalankan menggunakan pendekatan *parallel processing* melalui ThreadPoolExecutor dengan 40 workers. Setelah gold dataset berhasil dibangun dan dataset dibagi menjadi train, validation, dan fixed test set, langkah berikutnya adalah mengevaluasi performa baseline sentiment model. Pada tahap ini, prediksi sentimen yang dihasilkan oleh model baseline (tersimpan pada kolom baseline_pred) dibandingkan langsung dengan label final manusia (label) menggunakan beberapa metrik evaluasi standar dalam klasifikasi teks, yaitu:
- Precision – seberapa akurat model ketika memprediksi suatu kelas
- Recall – seberapa baik model menangkap seluruh instance dari kelas tersebut
- F1-score – keseimbangan antara precision dan recall
- Support – jumlah data pada setiap kelas

Evaluasi dilakukan pada dua subset utama:
- Validation Set → digunakan untuk memahami performa model sebelum eksperimen lanjutan seperti fine-tuning atau active learning
- Fixed Test Set → dataset evaluasi akhir yang dibekukan (frozen test set) untuk memastikan perbandingan performa model tetap konsisten di seluruh eksperimen

Selain menghasilkan classification report, proses evaluasi ini juga menyimpan beberapa artefak penting untuk keperluan analisis lebih lanjut, termasuk:
- Classification report (JSON & CSV) untuk analisis metrik model
- Distribusi label prediksi vs ground truth untuk melihat potensi bias model

Dengan melakukan evaluasi baseline ini, kita dapat memperoleh benchmark awal performa model, yang nantinya akan menjadi acuan ketika melakukan peningkatan model melalui LLM-assisted labeling refinement, fine-tuning, atau active learning iterations.

````python
# 3.2.4.1) Evaluasi baseline hybrid dengan menggunakan classification report dan confusion matrix. 
# Hasil evaluasi akan disimpan dalam format JSON, CSV, dan visualisasi PNG untuk dokumentasi.
labels_order = ["negative", "neutral", "positive"]

def evaluate_and_save(df_eval, split_name):
    y_true = df_eval["label"]
    y_pred = df_eval["baseline_pred"]

    print(f"\n=== BASELINE EVALUATION: {split_name.upper()} ===")
    print(f"Jumlah data evaluasi: {len(df_eval):,}")
    print(classification_report(y_true, y_pred, digits=4))

    # Save report json
    report_dict = classification_report(y_true, y_pred, digits=4, output_dict=True)
    with open(BASELINE_ARTIFACT_DIR / f"baseline_{split_name}_classification_report.json", "w", encoding="utf-8") as f:
        json.dump(report_dict, f, ensure_ascii=False, indent=2)

    # Save report csv
    report_df = pd.DataFrame(report_dict).transpose()
    report_df.to_csv(
        BASELINE_ARTIFACT_DIR / f"baseline_{split_name}_classification_report.csv",
        index=True,
        encoding="utf-8-sig"
    )

    # Save label distribution
    dist_df = pd.DataFrame({
        "y_true": y_true.value_counts(),
        "y_pred": y_pred.value_counts()
    }).fillna(0).astype(int)
    dist_df.to_csv(
        BASELINE_ARTIFACT_DIR / f"baseline_{split_name}_label_distribution.csv",
        encoding="utf-8-sig"
    )

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=labels_order)

    fig, ax = plt.subplots(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_order)
    disp.plot(values_format="d", ax=ax)
    ax.set_title(f"Confusion Matrix - Baseline Hybrid ({split_name})")
    fig.tight_layout()

    cm_png_path = BASELINE_ARTIFACT_DIR / f"baseline_{split_name}_confusion_matrix.png"
    fig.savefig(cm_png_path, dpi=200)
    plt.show()

    return report_dict


# Validation evaluation
val_report = evaluate_and_save(val_df, "validation")

# Fixed test evaluation (ini yang akan dipakai sampai akhir)
test_report = evaluate_and_save(test_df, "fixed_test")
````

````python
# 3.2.4.2) Menyimpan metadata run untuk dokumentasi dan audit trail. 
# Metadata ini mencakup informasi tentang dataset yang digunakan, ukuran split, urutan label, dan catatan penting terkait evaluasi baseline.
run_meta = {
    "run_id": RUN_ID,
    "timestamp": datetime.now().isoformat(),
    "labeled_done_csv": str(LABELED_DONE_PATH),
    "gold_dataset_csv": str(GOLD_DATASET_PATH),
    "train_csv": str(TRAIN_PATH),
    "validation_csv": str(VAL_PATH),
    "fixed_test_csv": str(FIXED_TEST_PATH),
    "train_size": int(len(train_df)),
    "validation_size": int(len(val_df)),
    "fixed_test_size": int(len(test_df)),
    "labels_order": labels_order,
    "notes": (
        "Baseline Hybrid evaluation with train/validation/fixed-test split. "
        "Fixed test set is frozen and must be reused for SetFit Round_0, "
        "Active Learning Round_1, and final model evaluation."
    )
}

with open(BASELINE_ARTIFACT_DIR / "run_meta.json", "w", encoding="utf-8") as f:
    json.dump(run_meta, f, ensure_ascii=False, indent=2)

print("\nSaved baseline artifacts:")
print("-", BASELINE_ARTIFACT_DIR / "sampling_meta.json")
print("-", BASELINE_ARTIFACT_DIR / "split_label_distribution.csv")
print("-", BASELINE_ARTIFACT_DIR / "baseline_validation_classification_report.json")
print("-", BASELINE_ARTIFACT_DIR / "baseline_validation_classification_report.csv")
print("-", BASELINE_ARTIFACT_DIR / "baseline_validation_confusion_matrix.png")
print("-", BASELINE_ARTIFACT_DIR / "baseline_fixed_test_classification_report.json")
print("-", BASELINE_ARTIFACT_DIR / "baseline_fixed_test_classification_report.csv")
print("-", BASELINE_ARTIFACT_DIR / "baseline_fixed_test_confusion_matrix.png")
print("-", BASELINE_ARTIFACT_DIR / "run_meta.json")
````

### **4. Fine Tuning**

#### 4.1 Setfit (Sentence Transformer Fine-tuning)

Untuk meningkatkan performa model tanpa membutuhkan ribuan data berlabel tambahan, digunakan pendekatan SetFit (Sentence Transformer Fine-Tuning) dengan base model:

`sentence-transformers/paraphrase-multilingual-mpnet-base-v2`

Pendekatan ini bekerja pada level kalimat, bukan hanya kata per kata, sehingga mampu menangkap:
- Makna utuh sebuah opini
- Nuansa ekspresi pengguna
- Perbedaan konteks dalam kalimat yang terlihat mirip

Keunggulan utama SetFit adalah kemampuannya meningkatkan akurasi secara signifikan meskipun hanya menggunakan dataset berlabel dalam jumlah terbatas.

Dengan metode ini, model menjadi lebih adaptif terhadap gaya bahasa reviewer gadget Indonesia tanpa memerlukan komputasi besar atau retraining penuh dari awal.

````python
# Libraries
import gc
import json
import os
import random
import time
from datetime import datetime
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
from datasets import Dataset
from setfit import SetFitModel, SetFitTrainer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
````

````python
# 4.1.0) Konfigurasi dan setup untuk fine-tuning SetFit dengan dataset hasil labeling manual.
PROJECT_ROOT = Path(".").resolve()

TRAIN_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "train_split.csv"
VAL_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "validation_split.csv"
FIXED_TEST_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "fixed_test_set.csv"

VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42

# tokenizer log
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# Artifacts
RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
FINETUNE_DIR = PROJECT_ROOT / "dungeon" / "setfit_finetune" / f"run_{RUN_ID}"
CHECKPOINT_DIR = FINETUNE_DIR / "checkpoints"
FINAL_DIR = FINETUNE_DIR / "final_model"

FINETUNE_DIR.mkdir(parents=True, exist_ok=True)
CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
FINAL_DIR.mkdir(parents=True, exist_ok=True)

print("Project root         :", PROJECT_ROOT)
print("Fine-tune output dir :", FINETUNE_DIR)

# reproducibility
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
````

Sebagai fondasi konfigurasi sebelum proses fine-tuning dimulai. Tujuan utamanya adalah memastikan proses training berjalan rapi, reproducible, dan mudah ditracking lewat artifacts.

Secara garis besar, langkah yang dilakukan adalah:
1. Menentukan Sumber Data Label Manual (Golden Dataset), Pipeline fine-tuning menggunakan baseline labeling sample yang sudah diberi label manual (`baseline_labeling_sample_done.csv`) sebagai golden dataset.
2. Menstandarkan Label & Aturan Validasi, Ditetapkan label yang dianggap valid hanya `negative`, `neutral`, `positive`. Ini memastikan training dan evaluasi tetap konsisten dan menghindari label “liar” akibat typo atau format data yang tidak seragam.
3. Menyiapkan Struktur Artifacts (dungeon/setfit_finetune) untuk Tracking Training Run, Ini penting untuk workflow yang “semi-production ready”, karena memudahkan audit, rollback, dan reproducibility. Sehingga setiap eksperimen fine-tuning memiliki jejak yang jelas, termasuk:
    - Checkpoints (model intermediate)
	- Final model (model terbaik)
	- Output evaluasi (metrics/report)

````python
# 4.1.1) Cek device yang tersedia dan dataset untuk memastikan semuanya siap sebelum proses fine-tuning dimulai.
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device yang dipakai  :", device)

gc.collect()
if device == "cuda":
    try:
        torch.cuda.empty_cache()
    except Exception:
        pass

# Load dataset splits
for p in [TRAIN_PATH, VAL_PATH, FIXED_TEST_PATH]:
    if not p.exists():
        raise FileNotFoundError(f"File split tidak ditemukan: {p}")

train_df = pd.read_csv(TRAIN_PATH)
val_df = pd.read_csv(VAL_PATH)
test_df = pd.read_csv(FIXED_TEST_PATH)

print(f"\nLoaded train set      : {len(train_df):,}")
print(f"Loaded validation set : {len(val_df):,}")
print(f"Loaded fixed test set : {len(test_df):,}")
````

````python
# 4.1.2) Validasi kolom yang diperlukan dan pre-processing sederhana untuk memastikan data siap digunakan untuk fine-tuning.
required_cols = ["text", "label"]
for split_name, df_split in {
    "train": train_df,
    "validation": val_df,
    "fixed_test": test_df
}.items():
    missing = [c for c in required_cols if c not in df_split.columns]
    if missing:
        raise ValueError(
            f"Kolom wajib pada {split_name} tidak lengkap. Missing: {missing}\n"
            f"Kolom tersedia: {list(df_split.columns)}"
        )

    df_split["text"] = df_split["text"].astype(str).str.strip()
    df_split["label"] = df_split["label"].astype(str).str.lower().str.strip()

# filter data kosong dan label tidak valid (jika ada)
before_train = len(train_df)
before_val = len(val_df)
before_test = len(test_df)

train_df = train_df[(train_df["text"] != "") & (train_df["label"].isin(VALID_LABELS))].copy()
val_df   = val_df[(val_df["text"] != "") & (val_df["label"].isin(VALID_LABELS))].copy()
test_df  = test_df[(test_df["text"] != "") & (test_df["label"].isin(VALID_LABELS))].copy()

print("\nFilter valid split:")
print(f"Train      -> sebelum: {before_train} | sesudah: {len(train_df)} | drop: {before_train-len(train_df)}")
print(f"Validation -> sebelum: {before_val} | sesudah: {len(val_df)} | drop: {before_val-len(val_df)}")
print(f"Fixed Test -> sebelum: {before_test} | sesudah: {len(test_df)} | drop: {before_test-len(test_df)}")

print("\nDistribusi label - Train:")
print(train_df["label"].value_counts(dropna=False))

print("\nDistribusi label - Validation:")
print(val_df["label"].value_counts(dropna=False))

print("\nDistribusi label - Fixed Test:")
print(test_df["label"].value_counts(dropna=False))

# save snapshot distribusi split
split_dist_df = pd.DataFrame({
    "train": train_df["label"].value_counts(),
    "validation": val_df["label"].value_counts(),
    "fixed_test": test_df["label"].value_counts()
}).fillna(0).astype(int)

split_dist_df.to_csv(FINETUNE_DIR / "split_label_distribution.csv", encoding="utf-8-sig")

# Convert ke Hugging Face Dataset untuk kompatibilitas dengan SetFitTrainer
train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True), preserve_index=False)
val_dataset = Dataset.from_pandas(val_df.reset_index(drop=True), preserve_index=False)
````

### Distribusi Data Labeled & Stratified Split (Total 1.071 Sampel)

Untuk menjamin objektivitas evaluasi dan stabilitas model, dataset berlabel (*Golden Dataset*) yang berjumlah 1.071 sampel dibagi menjadi tiga bagian menggunakan skema **60% Train / 20% Validation / 20% Fixed Test**.

Berikut adalah rincian distribusinya:

| Kelas      | Total Data | Persentase | Train (60%) | Validation (20%) | Fixed Test (20%) |
|------------|------------|------------|-------------|------------------|------------------|
| ⚪ Neutral | 659        | 61.53%     | 395         | 132              | 132              |
| 🔴 Negative| 318        | 29.69%     | 191         | 63               | 64               |
| 🟢 Positive| 94         | 8.78%      | 56          | 19               | 19               |
| **Total** | **1.071** | **100.00%**| **642** | **214** | **215** |

*(Catatan: Filter validasi menunjukkan 0 data yang di-drop, memastikan seluruh dataset dalam kondisi bersih dan siap pakai).*

#### Kenapa Menggunakan Skema 60/20/20?
Pembagian ini dipilih secara strategis untuk menyeimbangkan kemampuan belajar model dan validitas pengujian:
- **Train Set (642 data):** Porsi terbesar agar model punya cukup variasi kosakata untuk mempelajari pola bahasa gaul netizen.
- **Validation Set (214 data):** Digunakan di "balik layar" saat proses *training* berlangsung (misal untuk *early stopping* atau *hyperparameter tuning*) agar model tidak *overfitting*.
- **Fixed Test Set (215 data):** Ini adalah "benchmark Utama" yang di-freeze. Kumpulan data uji ini tidak akan pernah disentuh selama *training*. Tujuannya agar kita bisa membandingkan performa *Baseline Model* dengan model hasil *Active Learning* nanti secara adil (*apple-to-apple*).

#### Pentingnya "Stratified Split"
Pemisahan data di atas dilakukan secara **stratified** (berlapis) berdasarkan label sentimen.
Seperti yang terlihat pada tabel, kelas **Positive** adalah minoritas ekstrim (hanya ~8.7%). Jika kita membelah data secara acak (*random split*), ada risiko kelas positif tidak terwakili dengan baik di *Validation* atau *Test Set*. 

Dengan *stratification*, kita memaksa algoritma untuk mendistribusikan proporsi kelas secara merata. Hasilnya, kelas Positif tetap berada di kisaran ~8-9% baik di data Train, Val, maupun Test. Hal ini memastikan evaluasi metrik (seperti *Macro/Weighted F1-score*) tetap representatif dan tidak bias terhadap mayoritas kelas Netral.

````python
# 4.1.3) Load base model SetFit yang akan digunakan untuk fine-tuning.
BASE_MODEL = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"

print(f"\nLoading SetFit base model: {BASE_MODEL}")

model = SetFitModel.from_pretrained(
    BASE_MODEL,
    labels=VALID_LABELS
)

# pindahkan ke device
try:
    model.model_body.to(device)
except Exception:
    pass

if getattr(model, "model_head", None) is not None:
    try:
        model.model_head.to(device)
    except Exception:
        pass
````

````python
# 4.1.4) Train model dengan konfigurasi yang disesuaikan untuk GPU yang digunakan.
# Google Cloud GPU (NVIDIA L4)
if device == "cuda":
    NUM_ITERATIONS = 20
    BATCH_SIZE = 64
    NUM_EPOCHS = 3
else:
    NUM_ITERATIONS = 6
    BATCH_SIZE = 16
    NUM_EPOCHS = 2

trainer = SetFitTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    batch_size=BATCH_SIZE,
    num_epochs=NUM_EPOCHS,
    num_iterations=NUM_ITERATIONS,
)

# Fix kompatibilitas SetFit vs Transformers
if hasattr(trainer, "args"):
    if not hasattr(trainer.args, "eval_strategy") and hasattr(trainer.args, "evaluation_strategy"):
        trainer.args.eval_strategy = trainer.args.evaluation_strategy
````

#### Konfigurasi Fine-Tuning SetFit (GPU Optimized)

Dalam proses *fine-tuning* algoritma SetFit, konfigurasi diatur agar dinamis mengikuti ketersediaan perangkat keras. Mengingat eksperimen ini ditargetkan untuk berjalan di atas **Google Cloud Vertex AI dengan GPU NVIDIA L4 (24GB VRAM)**, parameter sengaja dimaksimalkan untuk memanfaatkan tenaga komputasi tersebut secara efisien.

Berikut adalah konfigurasi utama yang digunakan pada mode GPU (`device == "cuda"`):
- `NUM_ITERATIONS = 20`
- `BATCH_SIZE = 64`
- `NUM_EPOCHS = 3`

#### Rincian Parameter:

1. **`NUM_ITERATIONS = 20` (Contrastive Learning)**
   Parameter ini menentukan seberapa banyak kombinasi pasangan kalimat (*sentence pairs*) yang dibentuk secara acak untuk dipelajari oleh model. 
   - Karena GPU L4 sangat cepat, kita bisa mendorong iterasi ke angka 20 (jauh di atas standar awal 12). Semakin banyak iterasi, model akan semakin jeli dalam memahami *semantic similarity* (kemiripan makna) antar komentar bahasa gaul yang berisik.

2. **`BATCH_SIZE = 64` (Proses Data Sekaligus)**
   *Batch size* adalah jumlah data yang ditelan model dalam satu kali jalan sebelum ia memperbarui bobot pemahamannya. 
   - Angka 64 adalah *sweet spot* untuk kapasitas VRAM 24GB pada GPU L4. *Batch* yang besar membuat arah pembaruan gradien lebih stabil (tidak melompat-lompat) dan memangkas waktu *training* secara signifikan.

3. **`NUM_EPOCHS = 3` (Pencegahan Overfitting)**
   *Epoch* berarti berapa kali model disuruh membaca ulang seluruh data *training* (±642 sampel) dari awal hingga akhir.
   - Tiga putaran sudah sangat cukup untuk membuat model beradaptasi dengan konteks tanpa membuatnya sekadar "menghafal" kunci jawaban. Jika *epoch* disetel terlalu tinggi untuk dataset berukuran ratusan, model berisiko tinggi terkena *overfitting* (hanya jago di *train set*, tapi gagal di *test set*).

````python
# 4.1.5) Proses fine-tuning model dengan monitoring waktu untuk mendapatkan estimasi durasi training.
print("\nSedang melatih ulang model (SetFit Round_0)...")
print(
    f"Config -> device={device}, "
    f"batch_size={BATCH_SIZE}, "
    f"num_epochs={NUM_EPOCHS}, "
    f"num_iterations={NUM_ITERATIONS}"
)
print(f"Checkpoint dir: {CHECKPOINT_DIR}")
print(f"Final model dir: {FINAL_DIR}")

t0 = time.time()
trainer.train()
t1 = time.time()

train_minutes = (t1 - t0) / 60
print(f"\nTraining selesai. Durasi: {train_minutes:.2f} menit")
````

#### 4.2 Active Learning

##### Strategi Active Learning (Round 1): Fokus pada Titik Buta Model

Secara sederhana, **Active Learning** adalah pendekatan proaktif di mana model secara otomatis menyeleksi dan menyodorkan data yang paling "membingungkan" bagi dirinya, lalu meminta kita (manusia) untuk memberikan kunci jawaban yang benar. Alih-alih kita yang memilih data secara acak untuk dilabeli, model-lah yang "meminta bantuan" tepat di bagian yang ia rasa paling sulit.

##### Diperlukan Active Learning karena:
1. **Efisiensi Waktu & Biaya:** *Labeling* manual (terutama dengan bantuan LLM + verifikasi manusia) memakan banyak *resource*. Kita tidak ingin membuang waktu melabeli data yang sebenarnya sudah dipahami dengan baik oleh model.
2. **Optimalisasi Dataset Awal:** Basis data kita (*Gold Dataset* sejumlah 1.071 sampel) sudah cukup solid sebagai fondasi awal.
3. **Class Imbalance yang Ekstrim:** Kelas *Positive* sangat minoritas (kurang dari 10%). Menambah data secara *random* kemungkinan besar hanya akan menambah tumpukan data *Neutral* atau *Negative*.
4. **Fokus pada "Decision Boundary":** Model sudah pintar secara umum, kita hanya perlu menajamkan batas keputusannya di area yang masih abu-abu.

> Jika *Fine-tuning (Round 0)* ibarat **"latihan umum"** untuk memahami dasar-dasar bahasa gaul netizen, maka *Active Learning* adalah **"latihan khusus/remedial"**. Model tidak lagi disuruh belajar semua hal dari nol, melainkan dipaksa berlatih intensif khusus pada soal-soal yang ia sering jawab salah.

##### Evaluasi Round 0
Berdasarkan hasil *fine-tuning* sebelumnya:
- Kelas **Negative** dan **Neutral** sudah sangat kuat (*Precision & Recall* tinggi).
- Kelas **Positive** masih sangat lemah (*Recall* rendah). 
- Model terlihat cenderung **"main aman"** ketika ia ragu apakah sebuah pujian itu positif atau sarkas, ia akan membuangnya ke kelas *Neutral*.

Daripada menambah ribuan label baru secara acak (*random sampling*), kita memerlukan strategi ekstraksi data yang jauh lebih maksimal.

##### Tiga Strategi Active Learning (Round 1)
Untuk *Round 1* ini, kita akan menarik sampel data baru dari *Unlabeled Pool* menggunakan tiga kriteria spesifik:
1. **Uncertainty Sampling:** Memilih komentar-komentar di mana model memiliki tingkat kepercayaan diri (*confidence score*) yang paling rendah/ragu-ragu (misalnya probabilitas antar kelas beda tipis).
2. **Positive Mining:** Secara khusus memburu dan memprioritaskan kalimat-kalimat yang memiliki indikasi/sinyal kelas *Positive* untuk menyuntikkan lebih banyak contoh ke kelas minoritas ini.
3. **Disagreement Cases:** Mengambil kasus-kasus di mana *Baseline Model* (Hybrid) dan model SetFit *Round 0* memberikan tebakan yang berbeda. Perbedaan pendapat ini biasanya menandakan adanya kalimat yang sangat ambigu atau memiliki struktur sentimen ganda (*mixed sentiment*).

#### 4.2.1 Membuat Tambahan Dataset Untuk Proses Active Learning Round 1

````python
import gc
import hashlib
import json
import os
import random
import time
from datetime import datetime
from pathlib import Path

import numpy as np
import pandas as pd
import torch
from setfit import SetFitModel
````

````python
# 4.2.1.0) Konfigurasi dan setup untuk proses active learning round 1, termasuk paths ke dataset, model, dan direktori output untuk hasil seleksi sampel yang akan dilabeli.
PROJECT_ROOT = Path(".").resolve()

POOL_PATH = PROJECT_ROOT / "dungeon" / "processed_dataset" / "hybrid_sentiment_output.csv"
GOLD_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "gold_dataset_v1.csv"

SETFIT_ROOT = PROJECT_ROOT / "dungeon" / "setfit_finetune"
latest_run = sorted(SETFIT_ROOT.glob("run_*"))[-1]
MODEL_DIR = latest_run / "final_model"

VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# quota round_1
N_LABEL_ROUND1 = 300
N_UNC = 170
N_POS = 110
N_DIS = 20
assert N_UNC + N_POS + N_DIS == N_LABEL_ROUND1

RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
AL_DIR = PROJECT_ROOT / "dungeon" / "active_learning" / f"round_1_run_{RUN_ID}"
AL_DIR.mkdir(parents=True, exist_ok=True)

OUT_LABEL_PATH = AL_DIR / "to_label_round_1.csv"
META_PATH = AL_DIR / "round_1_selection_meta.json"
SCORED_POOL_PATH = AL_DIR / "scored_pool_round_1.csv"

print("Project root          :", PROJECT_ROOT)
print("Model dir             :", MODEL_DIR)
print("Pool path             :", POOL_PATH)
print("Gold path             :", GOLD_PATH)
print("Active learning dir   :", AL_DIR)
````

````python
# 4.2.1.1) Cek keberadaan model yang sudah di-fine-tune untuk memastikan proses active learning bisa dilanjutkan
if not MODEL_DIR.exists():
    raise FileNotFoundError(f"Model dir tidak ditemukan: {MODEL_DIR}")

print("\nLoading SetFit model Round_0...")
model = SetFitModel.from_pretrained(str(MODEL_DIR))
print("Model berhasil dimuat.")
````

````python
# 4.2.1.2) Load dataset pool dan gold, serta melakukan pre-processing dasar seperti pembersihan teks, validasi kolom yang diperlukan, dan anti-leakage untuk memastikan data pool siap digunakan untuk seleksi sampel active learning.
if not POOL_PATH.exists():
    raise FileNotFoundError(f"Pool file tidak ditemukan: {POOL_PATH}")

if not GOLD_PATH.exists():
    raise FileNotFoundError(f"Gold file tidak ditemukan: {GOLD_PATH}")

df_pool = pd.read_csv(POOL_PATH)
df_gold = pd.read_csv(GOLD_PATH)

print(f"\nPool loaded : {len(df_pool):,}")
print(f"Gold loaded : {len(df_gold):,}")

required_pool_cols = ["clean_text", "sentiment_label", "sentiment_score", "source_channel"]
missing = [c for c in required_pool_cols if c not in df_pool.columns]
if missing:
    raise ValueError(f"Pool missing columns: {missing}. Kolom tersedia: {list(df_pool.columns)}")

# clean minimal
df_pool["clean_text"] = df_pool["clean_text"].astype(str).str.strip()
df_pool["sentiment_label"] = df_pool["sentiment_label"].astype(str).str.lower().str.strip()
df_pool = df_pool[df_pool["clean_text"] != ""].copy()
df_pool = df_pool[df_pool["sentiment_label"].isin(VALID_LABELS)].copy()

# detect kolom text di gold
possible_text_cols = ["text", "clean_text", "comment", "raw_text"]
gold_text_col = next((c for c in possible_text_cols if c in df_gold.columns), None)
if gold_text_col is None:
    raise ValueError(f"Kolom text di gold tidak ketemu. Kolom tersedia: {list(df_gold.columns)}")

df_gold["gold_text"] = df_gold[gold_text_col].astype(str).str.strip()
df_gold = df_gold[df_gold["gold_text"] != ""].copy()

def hash_text(s: str) -> str:
    return hashlib.md5(s.encode("utf-8")).hexdigest()

df_pool["text_hash"] = df_pool["clean_text"].apply(hash_text)
df_gold["text_hash"] = df_gold["gold_text"].apply(hash_text)

# anti leakage: buang semua data yang sudah pernah masuk gold
gold_hashes = set(df_gold["text_hash"].unique().tolist())

before = len(df_pool)
df_pool = df_pool[~df_pool["text_hash"].isin(gold_hashes)].copy()
after = len(df_pool)

print(f"Anti-leakage exclude gold -> sebelum={before:,} | sesudah={after:,} | drop={before-after:,}")

# dedup pool by text hash (opsional tapi sangat membantu)
before_dedup = len(df_pool)
df_pool = df_pool.drop_duplicates(subset=["text_hash"]).copy()
after_dedup = len(df_pool)

print(f"Dedup pool by text_hash  -> sebelum={before_dedup:,} | sesudah={after_dedup:,} | drop={before_dedup-after_dedup:,}")
````

Sebelum pengambilan dataset baru, kita harus memastikan kolam kandidat data (*Unlabeled Pool*) dalam keadaan sangat steril. Log di atas menunjukkan proses pembersihan ketat untuk mencegah "kebocoran soal ujian" (*data leakage*).

Berikut rincian proses sanitasi yang terjadi:

1. **Initial Load:** Kita memulai dengan **4.722** data *unlabeled* (sebagai kolam kandidat) dan **1.071** data *Gold* (data yang sudah berlabel dan menjadi bagian dari *Train/Val/Test set*).
2. **Anti-Leakage Filter (Drop: 1.198 data):** Sistem secara agresif membuang semua komentar di dalam *Pool* yang *hash* teksnya identik dengan data di *Gold Dataset*. Angka *drop* yang sedikit lebih besar dari jumlah *Gold* (1.198 vs 1.071) membuktikan bahwa algoritma berhasil memberisihkan duplikat dari data *Gold* yang tersebar di dalam *Pool*. Langkah ini memastikan bahwa model tidak akan pernah dilatih menggunakan data yang sama dua kali, apalagi "mengintip" data dari *Test Set*.
3. **Internal Deduplication (Drop: 82 data):** Sistem mendeteksi dan menghapus 82 komentar duplikat yang ada di dalam *Pool* itu sendiri. Ini sangat penting agar kuota pelabelan manual kita nanti tidak terbuang sia-sia untuk melabeli komentar *spam* yang di-copy-paste berulang kali oleh netizen.

>Setelah melewati filter berlapis, kita kini memiliki **3.442 kandidat komentar** yang 100% baru, unik, dan belum pernah dilihat oleh model sebelumnya. Dari kolam steril inilah model akan mulai menyeleksi 300 data yang paling membingungkan baginya.

````python
# 4.2.1.3) Scoring pool dengan model SetFit untuk mendapatkan probabilitas prediksi, label prediksi, serta metrik ketidakpastian seperti entropy dan margin untuk setiap sampel di pool.
texts = df_pool["clean_text"].tolist()

print("\nScoring pool dengan SetFit predict_proba...")
t0 = time.time()
proba = model.predict_proba(texts)
proba = np.array(proba)
t1 = time.time()

print(f"Scoring selesai dalam {(t1 - t0)/60:.2f} menit")

# mapping label -> index
if hasattr(model, "labels"):
    model_labels = list(model.labels)
else:
    model_labels = VALID_LABELS

idx_map = {lab: model_labels.index(lab) for lab in VALID_LABELS}

p_neg = proba[:, idx_map["negative"]]
p_neu = proba[:, idx_map["neutral"]]
p_pos = proba[:, idx_map["positive"]]

pred_idx = proba.argmax(axis=1)
pred_label = [model_labels[i] for i in pred_idx]

# entropy
eps = 1e-12
entropy = -(proba * np.log(proba + eps)).sum(axis=1)

# margin
sorted_proba = np.sort(proba, axis=1)
margin = sorted_proba[:, -1] - sorted_proba[:, -2]

df_pool["pred_label_setfit"] = pred_label
df_pool["p_negative"] = p_neg
df_pool["p_neutral"] = p_neu
df_pool["p_positive"] = p_pos
df_pool["uncertainty_entropy"] = entropy
df_pool["uncertainty_margin"] = margin

df_pool.to_csv(SCORED_POOL_PATH, index=False, encoding="utf-8-sig")
print("Scored pool saved:", SCORED_POOL_PATH)
````

````python
# 4.2.1.4) Seleksi sampel untuk labeling dengan strategi kombinasi antara uncertainty sampling, positive mining, dan disagreement sampling untuk memastikan sampel yang dipilih memiliki nilai informasi yang tinggi untuk proses active learning

# A. Uncertainty
cand_unc = (
    df_pool
    .sort_values(["uncertainty_entropy", "uncertainty_margin"], ascending=[False, True])
    .head(N_UNC)
    .copy()
)
cand_unc["selection_reason"] = "uncertainty"

# B. Positive mining
cand_pos = df_pool.copy()
cand_pos["pos_rank_score"] = (
    0.75 * cand_pos["p_positive"] +
    0.25 * cand_pos["uncertainty_entropy"]
)

cand_pos = (
    cand_pos
    .loc[~cand_pos["text_hash"].isin(cand_unc["text_hash"])]
    .sort_values("pos_rank_score", ascending=False)
    .head(N_POS)
    .copy()
)
cand_pos["selection_reason"] = "positive_mining"

# C. Disagreement: baseline hybrid vs setfit berbeda
disagree = df_pool[
    (df_pool["sentiment_label"].isin(VALID_LABELS)) &
    (df_pool["sentiment_label"] != df_pool["pred_label_setfit"])
].copy()

disagree["setfit_conf"] = disagree[["p_negative", "p_neutral", "p_positive"]].max(axis=1)

used_hashes = set(pd.concat([cand_unc, cand_pos])["text_hash"].tolist())

disagree = (
    disagree
    .loc[~disagree["text_hash"].isin(used_hashes)]
    .sort_values(["setfit_conf", "uncertainty_entropy"], ascending=[False, False])
    .head(N_DIS)
    .copy()
)
disagree["selection_reason"] = "disagreement"

# gabung
df_sel = pd.concat([cand_unc, cand_pos, disagree], ignore_index=True)
df_sel = df_sel.drop_duplicates(subset=["text_hash"]).copy()

# kalau karena overlap jumlah kurang, isi dari uncertainty cadangan
if len(df_sel) < N_LABEL_ROUND1:
    need = N_LABEL_ROUND1 - len(df_sel)
    used_hashes = set(df_sel["text_hash"].tolist())

    fallback = (
        df_pool
        .loc[~df_pool["text_hash"].isin(used_hashes)]
        .sort_values(["uncertainty_entropy", "uncertainty_margin"], ascending=[False, True])
        .head(need)
        .copy()
    )
    fallback["selection_reason"] = "fallback_uncertainty"

    df_sel = pd.concat([df_sel, fallback], ignore_index=True)

df_sel = df_sel.sample(frac=1.0, random_state=SEED).reset_index(drop=True)

print("\nFinal selected:", len(df_sel))
print(df_sel["selection_reason"].value_counts(dropna=False))
````

##### Mekanisme Eksekusi Active Learning (Round 1)

Tahap ini adalah implementasi nyata dari strategi *Active Learning*. Kita tidak lagi memilih data secara acak untuk dilabeli secara manual, melainkan menggunakan algoritma untuk menambang (*mining*) komentar yang paling informatif dan krusial bagi perkembangan model.

Proses ekstraksi data ini dijalankan melalui empat langkah sistematis:

##### 1. Persiapan Sumber Data & Anti-Leakage
- **Unlabeled Pool:** Menggunakan dataset besar hasil prediksi *Baseline Model* (Hybrid) yang belum pernah disentuh oleh *annotator* manusia.
- **Data Sanitization (Anti-Leakage):** Sistem secara otomatis membandingkan *hash* teks dari *Unlabeled Pool* dengan *Gold Dataset* (1.071 data yang sudah dilabeli di fase sebelumnya). Teks yang sama persis atau duplikat akan dibuang (*drop*) untuk memastikan tidak ada data yang dilabeli dua kali (*zero data leakage*).

##### 2. Scoring Menggunakan Model Round 0
Model SetFit hasil *fine-tuning* putaran pertama (Round 0) ditugaskan kembali bukan untuk memprediksi hasil akhir, melainkan bertindak sebagai **"Penilai Tingkat Kesulitan"**. Model akan membaca seluruh sisa komentar yang belum berlabel dan mengeluarkan nilai *probability*, *entropy* (tingkat kebingungan), dan *margin* (selisih yakin antara prediksi juara 1 dan juara 2).

##### 3. Penentuan Target Labeling
Kita menetapkan kuota **300 data baru** untuk dilabeli pada Round 1 ini. Angka 300 dipilih sebagai *sweet spot* (titik seimbang) antara efisiensi waktu pelabelan manual manusia dan ekspektasi lonjakan performa model, karena data yang dipilih kali ini memiliki kualitas informasi (kepadatan konteks) yang jauh lebih tinggi daripada data acak.

##### 4. Strategic Selection (Pemilihan Berbasis Kuota Tetap)
Agar data tambahan tidak condong hanya pada satu jenis masalah, 300 data tersebut dibagi secara proporsional ke dalam tiga kantong strategi:
- **170 Data (Uncertainty Sampling):** Mencari komentar dengan nilai *Entropy* tertinggi. Ini adalah komentar-komentar "abu-abu" yang membuat model paling ragu-ragu. Tujuannya untuk mempertegas batas keputusan (*decision boundary*) model.
- **110 Data (Positive Mining):** Karena kelas Positif sangat minoritas (under-detected di Round 0), kita memaksa algoritma mencari komentar yang memiliki probabilitas ke arah positif, namun model belum berani mengeksekusinya. Tujuannya untuk memperkaya kosakata di kelas Positif.
- **20 Data (Disagreement Cases):** Memburu "Hard Cases" di mana prediksi *Baseline Hybrid* sangat bertolak belakang dengan prediksi *SetFit Round 0*. Perbedaan pendapat antar-mesin ini biasanya mengindikasikan struktur kalimat yang sarkas, ambigu, atau *mixed sentiment*.

````python
# 4.2.1.5) Export sampel yang sudah dipilih beserta metadata terkait ke file CSV untuk proses labeling manual, serta menyimpan metadata seleksi untuk dokumentasi dan audit trail.
out_cols = [
    "video_id", "author", "text", "clean_text", "source_channel",
    "sentiment_label", "sentiment_score",
    "pred_label_setfit", "p_negative", "p_neutral", "p_positive",
    "uncertainty_entropy", "uncertainty_margin",
    "selection_reason"
]

existing_cols = [c for c in out_cols if c in df_sel.columns]
df_out = df_sel[existing_cols].copy()

df_out["gold_label"] = ""
df_out["gold_notes"] = ""

df_out.to_csv(OUT_LABEL_PATH, index=False, encoding="utf-8-sig")
print("\nSaved to label file:", OUT_LABEL_PATH)
````

````python
# 4.2.1.6) Menyimpan metadata seleksi untuk dokumentasi dan audit trail, termasuk informasi tentang jumlah sampel yang dipilih, alasan seleksi, serta paths ke dataset yang digunakan dalam proses seleksi.
meta = {
    "round": 1,
    "run_id": RUN_ID,
    "seed": SEED,
    "n_label_round1": N_LABEL_ROUND1,
    "selection_mode": "fixed_quota",
    "quota": {
        "uncertainty": N_UNC,
        "positive_mining": N_POS,
        "disagreement": N_DIS
    },
    "paths": {
        "pool_path": str(POOL_PATH),
        "gold_path": str(GOLD_PATH),
        "model_dir": str(MODEL_DIR),
        "output_to_label_csv": str(OUT_LABEL_PATH),
        "scored_pool_csv": str(SCORED_POOL_PATH)
    },
    "counts": {
        "pool_after_antileakage": int(len(df_pool)),
        "selected_final": int(len(df_out))
    },
    "notes": (
        "Round_1 active learning selection. "
        "Gold dataset excluded to avoid leakage. "
        "Fixed test remains frozen because it is already included in gold_dataset_v1."
    )
}

with open(META_PATH, "w", encoding="utf-8") as f:
    json.dump(meta, f, ensure_ascii=False, indent=2)

print("Saved meta:", META_PATH)
````

#### 4.2.2 Run Active Learning Round 1

````python
import gc
import hashlib
import json
import os
import random
import time
from datetime import datetime
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
from datasets import Dataset
from setfit import SetFitModel, SetFitTrainer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
````

````python
# 4.2.2.0) Konfigurasi
PROJECT_ROOT = Path(".").resolve()

VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42

TRAIN_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "train_split.csv"
VAL_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "validation_split.csv"
FIXED_TEST_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "fixed_test_set.csv"

AL_ROOT = PROJECT_ROOT / "dungeon" / "active_learning"
latest_round1_dir = sorted(AL_ROOT.glob("round_1_run_*"))[-1]
ROUND1_DONE_PATH = latest_round1_dir / "round_1_labels_done.csv"

RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
FINETUNE_DIR = PROJECT_ROOT / "dungeon" / "setfit_finetune" / f"run_round1_{RUN_ID}"
CHECKPOINT_DIR = FINETUNE_DIR / "checkpoints"
FINAL_DIR = FINETUNE_DIR / "final_model"

FINETUNE_DIR.mkdir(parents=True, exist_ok=True)
CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
FINAL_DIR.mkdir(parents=True, exist_ok=True)

TRAIN_V1_PATH = latest_round1_dir / "train_split_round_1_merged.csv"

BASE_MODEL = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"

os.environ["TOKENIZERS_PARALLELISM"] = "false"

print("Project root            :", PROJECT_ROOT)
print("Round_1 active learning :", latest_round1_dir)
print("Round_1 labeled done    :", ROUND1_DONE_PATH)
print("Round_1 finetune dir    :", FINETUNE_DIR)

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
````

````python
# 4.2.2.1) Cek Device
device = "cuda" if torch.cuda.is_available() else "cpu"
print("\nDevice yang dipakai:", device)

gc.collect()
if device == "cuda":
    try:
        torch.cuda.empty_cache()
    except Exception:
        pass
````

````python
# 4.2.2.2) Load dataset
for p in [TRAIN_PATH, VAL_PATH, FIXED_TEST_PATH, ROUND1_DONE_PATH]:
    if not p.exists():
        raise FileNotFoundError(f"File tidak ditemukan: {p}")

train_df = pd.read_csv(TRAIN_PATH)
val_df = pd.read_csv(VAL_PATH)
test_df = pd.read_csv(FIXED_TEST_PATH)
df_r1 = pd.read_csv(ROUND1_DONE_PATH)

print(f"\nLoaded train_split   : {len(train_df):,}")
print(f"Loaded validation    : {len(val_df):,}")
print(f"Loaded fixed_test    : {len(test_df):,}")
print(f"Loaded round_1 done  : {len(df_r1):,}")
````

````python
# 4.2.2.3) Validasi kolom dan pre-processing dasar
possible_text_cols = ["text", "clean_text", "comment", "raw_text"]
possible_label_cols = ["gold_label", "label", "sentiment_label", "sentiment"]

def normalize_df(df, df_name="df"):
    text_col = next((c for c in possible_text_cols if c in df.columns), None)
    label_col = next((c for c in possible_label_cols if c in df.columns), None)

    if text_col is None or label_col is None:
        raise ValueError(
            f"[{df_name}] Kolom text/label tidak ketemu.\n"
            f"Kolom tersedia: {list(df.columns)}"
        )

    out = df[[text_col, label_col]].copy()
    out = out.rename(columns={text_col: "text", label_col: "label"})
    out["text"] = out["text"].astype(str).str.strip()
    out["label"] = out["label"].astype(str).str.lower().str.strip()
    out = out[(out["text"] != "") & (out["label"].isin(VALID_LABELS))].copy()
    return out

train_n = normalize_df(train_df, "train_split")
val_n = normalize_df(val_df, "validation_split")
test_n = normalize_df(test_df, "fixed_test_set")
r1_n = normalize_df(df_r1, "round_1_labels_done")

print("\nAfter normalize/filter:")
print("train usable      :", len(train_n))
print("validation usable :", len(val_n))
print("fixed_test usable :", len(test_n))
print("round_1 usable    :", len(r1_n))
````

````python
# 4.2.2.4) Menggabungkan dataset round_1 dengan train_split untuk membuat train_v1, dengan memastikan tidak ada kebocoran data antara round_1, validation, dan fixed test.
def hash_text(s: str) -> str:
    return hashlib.md5(s.encode("utf-8")).hexdigest()

train_n["text_hash"] = train_n["text"].apply(hash_text)
val_n["text_hash"] = val_n["text"].apply(hash_text)
test_n["text_hash"] = test_n["text"].apply(hash_text)
r1_n["text_hash"] = r1_n["text"].apply(hash_text)

# Jangan sampai round_1 masuk validation / fixed test
forbidden_hashes = set(val_n["text_hash"]).union(set(test_n["text_hash"]))

before_r1 = len(r1_n)
r1_n = r1_n[~r1_n["text_hash"].isin(forbidden_hashes)].copy()
after_r1 = len(r1_n)

print(f"\nRound_1 anti-leakage vs val/test -> sebelum={before_r1:,} | sesudah={after_r1:,} | drop={before_r1-after_r1:,}")

# merge ke train
train_v1 = pd.concat([train_n, r1_n], ignore_index=True)
train_v1 = train_v1.drop_duplicates(subset=["text_hash"], keep="last").copy()
train_v1 = train_v1.drop(columns=["text_hash"])
val_n = val_n.drop(columns=["text_hash"])
test_n = test_n.drop(columns=["text_hash"])

train_v1.to_csv(TRAIN_V1_PATH, index=False, encoding="utf-8-sig")
print("Merged train_v1 saved:", TRAIN_V1_PATH)
print("train_v1 size:", len(train_v1))

# save distribution snapshot
split_dist = pd.DataFrame({
    "train_v1": train_v1["label"].value_counts(),
    "validation": val_n["label"].value_counts(),
    "fixed_test": test_n["label"].value_counts()
}).fillna(0).astype(int)

split_dist.to_csv(FINETUNE_DIR / "split_label_distribution_round1.csv", encoding="utf-8-sig")

print("\nDistribusi label:")
print(split_dist)
````

##### Data Active Learning & Distribusi Baru (Round 1)

Setelah 300 data dari *Active Learning* dilabeli secara manual, langkah selanjutnya adalah gabungkan kembali ke dalam *Training Set* lama. Sebelum digabung, sistem melakukan pengecekan **Anti-Leakage** secara ketat terhadap *Validation* dan *Fixed Test Set*. Hasilnya: **0 data ter-drop**, yang membuktikan bahwa strategi pemilihan data kita sangat bersih dan tidak ada satupun "soal ujian" yang bocor ke materi latihan model.

Berikut adalah wajah baru dari distribusi data kita (`train_v1`):

**Key Takeaways (Dampak Strategis):**

````python
# 4.2.2.5) Convert ke Hugging Face Dataset untuk kompatibilitas dengan SetFitTrainer
train_dataset = Dataset.from_pandas(train_v1.reset_index(drop=True), preserve_index=False)
val_dataset = Dataset.from_pandas(val_n.reset_index(drop=True), preserve_index=False)

# LOAD MODEL
print(f"\nLoading SetFit base model: {BASE_MODEL}")

model = SetFitModel.from_pretrained(
    BASE_MODEL,
    labels=VALID_LABELS
)

try:
    model.model_body.to(device)
except Exception:
    pass

if getattr(model, "model_head", None) is not None:
    try:
        model.model_head.to(device)
    except Exception:
        pass
````

````python
# 4.2.2.6) Konfigurasi Trainer dan Train model dengan konfigurasi yang disesuaikan untuk GPU yang digunakan
if device == "cuda":
    NUM_ITERATIONS = 20
    BATCH_SIZE = 64
    NUM_EPOCHS = 3
else:
    NUM_ITERATIONS = 6
    BATCH_SIZE = 16
    NUM_EPOCHS = 2

trainer = SetFitTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    batch_size=BATCH_SIZE,
    num_epochs=NUM_EPOCHS,
    num_iterations=NUM_ITERATIONS,
)

# fix kompatibilitas beberapa versi transformers/setfit
if hasattr(trainer, "args"):
    if not hasattr(trainer.args, "eval_strategy") and hasattr(trainer.args, "evaluation_strategy"):
        trainer.args.eval_strategy = trainer.args.evaluation_strategy
````

````python
# 4.2.2.7) Proses fine-tuning model dengan monitoring waktu untuk mendapatkan estimasi durasi training
print("\nTraining SetFit Round_1...")
print(
    f"Config -> device={device}, "
    f"batch_size={BATCH_SIZE}, "
    f"num_epochs={NUM_EPOCHS}, "
    f"num_iterations={NUM_ITERATIONS}"
)
print("Checkpoint dir:", CHECKPOINT_DIR)
print("Final model dir:", FINAL_DIR)

t0 = time.time()
trainer.train()
t1 = time.time()

train_minutes = (t1 - t0) / 60
print(f"\nTraining selesai. Durasi: {train_minutes:.2f} menit")
````

````python
# 4.2.2.8) Evaluasi model pada validation set untuk mendapatkan metrik akurasi dan classification report.
y_true_val = val_n["label"].tolist()
y_pred_val = trainer.model.predict(val_n["text"].tolist())

val_acc = accuracy_score(y_true_val, y_pred_val)
val_report_dict = classification_report(y_true_val, y_pred_val, digits=4, output_dict=True)

metrics = {
    "run_id": RUN_ID,
    "round": 1,
    "device": device,
    "base_model": BASE_MODEL,
    "seed": SEED,
    "train_v1_size": int(len(train_v1)),
    "validation_size": int(len(val_n)),
    "fixed_test_size": int(len(test_n)),
    "batch_size": BATCH_SIZE,
    "num_epochs": NUM_EPOCHS,
    "num_iterations": NUM_ITERATIONS,
    "validation_accuracy": float(val_acc),
    "train_duration_minutes": float(train_minutes),
    "notes": "Round_1 retrain uses original frozen validation and frozen fixed test."
}

with open(FINETUNE_DIR / "metrics_validation_round1.json", "w", encoding="utf-8") as f:
    json.dump(
        {"metrics": metrics, "classification_report": val_report_dict},
        f,
        ensure_ascii=False,
        indent=2
    )

val_report_df = pd.DataFrame(val_report_dict).transpose()
val_report_df.to_csv(FINETUNE_DIR / "classification_report_validation_round1.csv", index=True, encoding="utf-8-sig")

print("\n=== VALIDATION EVALUATION (Round_1) ===")
print(val_report_df)

labels_order = ["negative", "neutral", "positive"]
cm_val = confusion_matrix(y_true_val, y_pred_val, labels=labels_order)

fig_val, ax_val = plt.subplots(figsize=(6, 5))
disp_val = ConfusionMatrixDisplay(cm_val, display_labels=labels_order)
disp_val.plot(values_format="d", ax=ax_val)
ax_val.set_title("Confusion Matrix - SetFit Round_1 (Validation)")
fig_val.tight_layout()

cm_val_path = FINETUNE_DIR / "confusion_matrix_validation_round1.png"
fig_val.savefig(cm_val_path, dpi=200)
plt.show()

print("\nValidation confusion matrix saved to:", cm_val_path)
````

#### 4.3 Fixed Test Set Evaluation (Baseline, Finetune R0 dan Active Learning R1)

`Fixed Test Set` adalah sekumpulan data uji yang sengaja dipisahkan sejak awal proses (*di-freeze*) dan diharamkan untuk disentuh selama fase *training* maupun *Active Learning*. Analogi sederhananya: ini adalah **"soal ujian standar"** yang kerahasiaannya dijaga ketat agar kita bisa mengukur perkembangan kecerdasan model (dari Baseline, Round 0 dan ke Round 1) secara konsisten.

#### 4.3.1 Analisis perbandingan model Base, Round_0, dan Round_1.

````python
import json
from datetime import datetime
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
from setfit import SetFitModel
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)
````

````python
# 4.3.1.0) Konfigurasi
PROJECT_ROOT = Path(".").resolve()

FIXED_TEST_PATH = PROJECT_ROOT / "dungeon" / "manual_labeling" / "fixed_test_set.csv"

SETFIT_ROOT = PROJECT_ROOT / "dungeon" / "setfit_finetune"
ROUND0_RUNS = sorted(SETFIT_ROOT.glob("run_[0-9]*"))
ROUND1_RUNS = sorted(SETFIT_ROOT.glob("run_round1_*"))

if not ROUND0_RUNS:
    raise FileNotFoundError("Folder run Round_0 tidak ditemukan di dungeon/setfit_finetune")
if not ROUND1_RUNS:
    raise FileNotFoundError("Folder run Round_1 tidak ditemukan di dungeon/setfit_finetune")

ROUND0_DIR = ROUND0_RUNS[-1]
ROUND1_DIR = ROUND1_RUNS[-1]

ROUND0_MODEL_DIR = ROUND0_DIR / "final_model"
ROUND1_MODEL_DIR = ROUND1_DIR / "final_model"

RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
SUMMARY_DIR = PROJECT_ROOT / "dungeon" / "summary_analysis" / f"fixed_test_comparison_run_{RUN_ID}"
SUMMARY_DIR.mkdir(parents=True, exist_ok=True)

VALID_LABELS = ["negative", "neutral", "positive"]
LABELS_ORDER = ["negative", "neutral", "positive"]

print("Project root       :", PROJECT_ROOT)
print("Fixed test path    :", FIXED_TEST_PATH)
print("Round_0 model dir  :", ROUND0_MODEL_DIR)
print("Round_1 model dir  :", ROUND1_MODEL_DIR)
print("Summary output dir :", SUMMARY_DIR)
````

````python
# 4.3.1.1) Load dataset fixed test, lakukan validasi kolom, pre-processing dasar, dan pastikan distribusi label sesuai untuk memastikan evaluasi perbandingan model berjalan dengan data yang bersih dan valid.
if not FIXED_TEST_PATH.exists():
    raise FileNotFoundError(f"Fixed test set tidak ditemukan: {FIXED_TEST_PATH}")

df_test = pd.read_csv(FIXED_TEST_PATH).copy()

required_cols = ["text", "label", "baseline_pred"]
missing = [c for c in required_cols if c not in df_test.columns]
if missing:
    raise ValueError(
        f"Kolom wajib fixed test tidak lengkap. Missing: {missing}\n"
        f"Kolom tersedia: {list(df_test.columns)}"
    )

df_test["text"] = df_test["text"].astype(str).str.strip()
df_test["label"] = df_test["label"].astype(str).str.lower().str.strip()
df_test["baseline_pred"] = df_test["baseline_pred"].astype(str).str.lower().str.strip()

before = len(df_test)
df_test = df_test[
    (df_test["text"] != "") &
    (df_test["label"].isin(VALID_LABELS)) &
    (df_test["baseline_pred"].isin(VALID_LABELS))
].copy()
after = len(df_test)

print(f"\nFixed test loaded -> sebelum: {before:,} | sesudah valid: {after:,} | drop: {before-after:,}")
print("\nDistribusi fixed test:")
print(df_test["label"].value_counts(dropna=False))

y_true = df_test["label"].tolist()
texts_test = df_test["text"].tolist()
````

````python
# 4.3.1.2) Load model Base, Round_0, dan Round_1 untuk melakukan evaluasi perbandingan pada fixed test set.
if not ROUND0_MODEL_DIR.exists():
    raise FileNotFoundError(f"Final model Round_0 tidak ditemukan: {ROUND0_MODEL_DIR}")
if not ROUND1_MODEL_DIR.exists():
    raise FileNotFoundError(f"Final model Round_1 tidak ditemukan: {ROUND1_MODEL_DIR}")

print("\nLoading SetFit Round_0 model...")
model_r0 = SetFitModel.from_pretrained(str(ROUND0_MODEL_DIR))

print("Loading SetFit Round_1 model...")
model_r1 = SetFitModel.from_pretrained(str(ROUND1_MODEL_DIR))
````

````python
# 4.3.1.3) Predict pada fixed test set dengan ketiga model (Baseline, Round_0, Round_1) untuk mendapatkan prediksi label yang akan digunakan.
print("\nPredicting on fixed test set...")

y_pred_baseline = df_test["baseline_pred"].tolist()
y_pred_r0 = model_r0.predict(texts_test)
y_pred_r1 = model_r1.predict(texts_test)
````

````python
# 4.3.1.4) Evaluasi performa ketiga model dengan metrik akurasi, F1-score, classification report, dan confusion matrix untuk mendapatkan gambaran lengkap tentang perbandingan performa model pada fixed test set.
def evaluate_model(y_true, y_pred, model_name, output_dir, labels_order):
    acc = accuracy_score(y_true, y_pred)
    macro_f1 = f1_score(y_true, y_pred, average="macro")
    weighted_f1 = f1_score(y_true, y_pred, average="weighted")

    report_dict = classification_report(
        y_true,
        y_pred,
        labels=labels_order,
        digits=4,
        output_dict=True
    )

    report_df = pd.DataFrame(report_dict).transpose()
    report_csv_path = output_dir / f"{model_name}_classification_report.csv"
    report_json_path = output_dir / f"{model_name}_classification_report.json"

    report_df.to_csv(report_csv_path, encoding="utf-8-sig")
    with open(report_json_path, "w", encoding="utf-8") as f:
        json.dump(report_dict, f, ensure_ascii=False, indent=2)

    cm = confusion_matrix(y_true, y_pred, labels=labels_order)

    fig, ax = plt.subplots(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_order)
    disp.plot(values_format="d", ax=ax)
    ax.set_title(f"Confusion Matrix - {model_name}")
    fig.tight_layout()

    cm_path = output_dir / f"{model_name}_confusion_matrix.png"
    fig.savefig(cm_path, dpi=200)
    plt.show()

    return {
        "model": model_name,
        "accuracy": float(acc),
        "macro_f1": float(macro_f1),
        "weighted_f1": float(weighted_f1),
        "report_csv": str(report_csv_path),
        "report_json": str(report_json_path),
        "confusion_matrix_png": str(cm_path),
    }
````

````python
# 4.3.1.5) Menampilkan classification report untuk ketiga model, metrik evaluasi serta confusion matrix.
print("\n=== BASELINE HYBRID (FIXED TEST) ===")
print(classification_report(y_true, y_pred_baseline, labels=LABELS_ORDER, digits=4))

baseline_metrics = evaluate_model(
    y_true, y_pred_baseline,
    model_name="baseline_hybrid_fixed_test",
    output_dir=SUMMARY_DIR,
    labels_order=LABELS_ORDER
)

print("\n=== SETFIT ROUND_0 (FIXED TEST) ===")
print(classification_report(y_true, y_pred_r0, labels=LABELS_ORDER, digits=4))

round0_metrics = evaluate_model(
    y_true, y_pred_r0,
    model_name="setfit_round0_fixed_test",
    output_dir=SUMMARY_DIR,
    labels_order=LABELS_ORDER
)

print("\n=== SETFIT ROUND_1 (FIXED TEST) ===")
print(classification_report(y_true, y_pred_r1, labels=LABELS_ORDER, digits=4))

round1_metrics = evaluate_model(
    y_true, y_pred_r1,
    model_name="setfit_round1_fixed_test",
    output_dir=SUMMARY_DIR,
    labels_order=LABELS_ORDER
)
````

````python
# 4.5.0) Konfigurasi
PROJECT_ROOT = Path(".").resolve()

INPUT_PATH = PROJECT_ROOT / "dungeon" / "cleaned_dataset" / "cleaned_data.csv"
PROCESSED_DIR = PROJECT_ROOT / "dungeon" / "processed_dataset"
FINAL_MODEL_DIR = PROJECT_ROOT / "dungeon" / "final_model" / "sentiment_model_v1"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

OUTPUT_PATH = PROCESSED_DIR / "sentiment_analyzed_setfit_final.csv"

VALID_LABELS = ["negative", "neutral", "positive"]

print("Project root     :", PROJECT_ROOT)
print("Input dataset    :", INPUT_PATH)
print("Expected model   :", FINAL_MODEL_DIR)
print("Output dataset   :", OUTPUT_PATH)
````

````python
# 4.5.1) Load Cleaned Dataset
if not INPUT_PATH.exists():
    raise FileNotFoundError(f"File cleaned dataset tidak ditemukan: {INPUT_PATH}")

df = pd.read_csv(INPUT_PATH).copy()

if "clean_text" not in df.columns:
    raise ValueError(f"Kolom 'clean_text' tidak ditemukan. Kolom tersedia: {list(df.columns)}")

df["clean_text"] = df["clean_text"].astype(str).str.strip()

before = len(df)
df = df[df["clean_text"] != ""].copy().reset_index(drop=True)
after = len(df)

print(f"\nLoaded cleaned dataset -> sebelum: {before:,} | sesudah valid: {after:,} | drop: {before-after:,}")
````

````python
# 4.5.2) Load Final Model
model_dir = None

if FINAL_MODEL_DIR.exists():
    model_dir = FINAL_MODEL_DIR
else:
    # fallback: ambil latest round_1 run
    setfit_root = PROJECT_ROOT / "dungeon" / "setfit_finetune"
    round1_runs = sorted(setfit_root.glob("run_round1_*"))
    if not round1_runs:
        raise FileNotFoundError(
            "Model final tidak ditemukan di dungeon/final_model/sentiment_model_v1 "
            "dan folder run_round1_* juga tidak ditemukan."
        )
    model_dir = round1_runs[-1] / "final_model"

if not model_dir.exists():
    raise FileNotFoundError(f"Model dir tidak ditemukan: {model_dir}")

print("\nLoading final SetFit model...")
print("Model path:", model_dir)

model = SetFitModel.from_pretrained(str(model_dir))
print("Model berhasil dimuat.")
````

````python
# 4.5.3) Prediksi sentimen untuk seluruh dataset yang sudah dibersihkan menggunakan model final yang sudah difreeze untuk mendapatkan label sentimen prediksi pada setiap komentar di dataset.
texts = df["clean_text"].tolist()

print(f"\nPredicting sentiment untuk {len(texts):,} komentar...")
t0 = time.time()

# label prediction
pred_labels = model.predict(texts)

# probability prediction
proba = model.predict_proba(texts)
proba = np.array(proba)

t1 = time.time()
infer_minutes = (t1 - t0) / 60

print(f"Prediction selesai dalam {infer_minutes:.2f} menit")
````

````python
# 4.5.4) Maping Label dan Probabilitas ke Data Frame.
if hasattr(model, "labels"):
    model_labels = list(model.labels)
else:
    model_labels = VALID_LABELS

# pastikan semua label target ada
missing_labels = [lab for lab in VALID_LABELS if lab not in model_labels]
if missing_labels:
    raise ValueError(
        f"Label model tidak lengkap. Missing: {missing_labels}. "
        f"Label tersedia di model: {model_labels}"
    )

idx_map = {lab: model_labels.index(lab) for lab in VALID_LABELS}

df["sentiment_label_final"] = pred_labels
df["p_negative"] = proba[:, idx_map["negative"]]
df["p_neutral"] = proba[:, idx_map["neutral"]]
df["p_positive"] = proba[:, idx_map["positive"]]

# skor final = probabilitas tertinggi dari label prediksi
df["sentiment_score_final"] = df[["p_negative", "p_neutral", "p_positive"]].max(axis=1)

# rapikan label
df["sentiment_label_final"] = df["sentiment_label_final"].astype(str).str.lower().str.strip()

# 5) SIMPLE SUMMARY
label_counts = df["sentiment_label_final"].value_counts()
label_pct = (df["sentiment_label_final"].value_counts(normalize=True) * 100).round(2)

summary_df = pd.DataFrame({
    "count": label_counts,
    "percentage (%)": label_pct
})

print("\nDistribusi sentiment_label_final:")
print(summary_df)
````

### **5. Aspect-Based Categorization (Otomatisasi Topik)**

Setelah model final SetFit selesai dan performanya tervalidasi menggunakan fixed test set, langkah berikutnya  adalah mengidentifikasi topik utama yang mendorong sentimen tersebut. 

Melalui `Aspect-Based Categorization`, komentar pengguna secara otomatis dikelompokkan ke dalam aspek produk seperti `kamera`, `baterai`, `performa` dll, sehingga analisis tidak hanya menjawab “bagaimana persepsi pasar”, tetapi juga “terhadap fitur apa persepsi tersebut terbentuk”.

Jika sentiment analysis menjawab emosi pasar, maka aspect-based categorization menjawab topik pembicaraan pasar. Tanpa aspek, kita hanya tahu 30% negatif. Dengan aspek, kita tahu 30% negatif itu karena kamera, baterai, atau software.

Tahap ini bertujuan untuk:
- Mengotomatisasi pengelompokan topik komentar
- Menghubungkan sentimen dengan aspek produk
- Mengubah data opini menjadi insight yang bisa ditindaklanjuti (actionable)

````python
# 5.3) Membuat regex pattern untuk setiap aspek berdasarkan keyword yang sudah didefinisikan, dengan mempertimbangkan variasi penulisan dan memastikan pencocokan yang akurat untuk mengidentifikasi aspek yang dibahas dalam komentar.
ASPECT_PATTERNS = {}

for aspect, keywords in ASPECTS_DICT.items():
    patterns = []
    seen = set()

    for kw in keywords:
        kw = kw.strip().lower()
        if not kw or kw in seen:
            continue

        seen.add(kw)

        if len(kw) <= 2:
            patterns.append(re.compile(rf"\b{re.escape(kw)}\b"))
        else:
            patterns.append(re.compile(rf"(?<!\w){re.escape(kw)}(?!\w)"))

    ASPECT_PATTERNS[aspect] = patterns

# 5.4) Tagging aspek untuk setiap komentar dengan memeriksa keberadaan keyword terkait aspek dalam teks komentar, dan menyimpan hasil tagging dalam format list untuk setiap baris di dataset.
def tag_aspects(text: str):
    if not isinstance(text, str) or not text.strip():
        return ["umum"]

    t = text.lower()
    tags = []

    for aspect, patterns in ASPECT_PATTERNS.items():
        if any(p.search(t) for p in patterns):
            tags.append(aspect)

    return tags if tags else ["umum"]

# 5.5) Tagging aspek untuk seluruh dataset dan melakukan explode pada kolom aspek untuk mendapatkan satu baris per aspek yang terdeteksi, sehingga memudahkan analisis aspek berbasis sentimen di langkah selanjutnya.
print("\nMenjalankan aspect tagging...")

df["aspects"] = df["clean_text"].apply(tag_aspects)
df_exploded = df.explode("aspects").copy()

print(f"Total row setelah explode aspek: {len(df_exploded):,}")

exploded_path = OUTPUT_DIR / "aspect_tagged_exploded.csv"
df_exploded.to_csv(exploded_path, index=False, encoding="utf-8-sig")
````

````python
# 5.7) Analisis aspek utama: Melakukan analisis lebih mendalam pada komentar yang menyebutkan aspek spesifik (tidak termasuk "umum") untuk memahami sentimen yang terkait dengan setiap aspek utama, serta mengidentifikasi aspek mana yang paling banyak dibahas dan bagaimana sentimen tersebar di setiap aspek tersebut.
df_analysis = df_exploded[df_exploded["aspects"] != "umum"].copy()

print(f"\nKomentar untuk analisis aspek utama (tanpa 'umum'): {len(df_analysis):,}")

if df_analysis.empty:
    raise ValueError("Semua komentar masuk kategori 'umum'. Tidak ada aspek spesifik untuk dianalisis.")
````

````python
# 5.8) Summary aspek tanpa z-score: Membuat summary table untuk setiap aspek utama yang mencakup jumlah total mention, distribusi sentimen (jumlah dan persentase untuk masing-masing label sentimen), rata-rata skor sentimen (polarity), serta memberikan status sentimen keseluruhan untuk setiap aspek berdasarkan rata-rata skor sentimen tersebut, tanpa melakukan normalisasi menggunakan z-score.
aspect_summary_raw = (
    df_analysis
    .groupby("aspects", dropna=False)
    .agg(
        total_mentions=("clean_text", "count"),
        negative_count=(LABEL_COL_FINAL, lambda x: (x == "negative").sum()),
        neutral_count=(LABEL_COL_FINAL, lambda x: (x == "neutral").sum()),
        positive_count=(LABEL_COL_FINAL, lambda x: (x == "positive").sum()),
        avg_sentiment=("polarity", "mean"),
    )
    .sort_values("total_mentions", ascending=False)
)

aspect_summary_raw["negative_pct"] = (
    aspect_summary_raw["negative_count"] / aspect_summary_raw["total_mentions"] * 100
).round(2)

aspect_summary_raw["neutral_pct"] = (
    aspect_summary_raw["neutral_count"] / aspect_summary_raw["total_mentions"] * 100
).round(2)

aspect_summary_raw["positive_pct"] = (
    aspect_summary_raw["positive_count"] / aspect_summary_raw["total_mentions"] * 100
).round(2)

def sentiment_status_raw(avg):
    if avg > 0.10:
        return "🟢 Positif"
    elif avg < -0.10:
        return "🔴 Negatif"
    else:
        return "🟡 Netral"

aspect_summary_raw["sentiment_status_raw"] = aspect_summary_raw["avg_sentiment"].apply(sentiment_status_raw)

raw_summary_path = OUTPUT_DIR / "aspect_summary_raw.csv"
aspect_summary_raw.to_csv(raw_summary_path, encoding="utf-8-sig")

print("\nSUMMARY ASPEK TANPA Z-SCORE (tanpa 'umum'):")
display(aspect_summary_raw)
````

````python
# 5.9) Summary aspek dengan z-score: Melakukan normalisasi rata-rata skor sentimen untuk setiap aspek menggunakan z-score untuk mengidentifikasi aspek mana yang memiliki sentimen relatif paling positif atau negatif dibandingkan dengan aspek lainnya, serta memberikan status sentimen berdasarkan z-score tersebut untuk memberikan insight yang lebih mendalam tentang aspek mana yang paling menonjol dalam hal sentimen positif atau negatif. 
# Aspek dengan z-score tinggi menunjukkan sentimen yang lebih positif dibandingkan rata-rata, sedangkan z-score rendah menunjukkan sentimen yang lebih negatif.
aspect_summary_z = aspect_summary_raw.copy()

mean_val = aspect_summary_z["avg_sentiment"].mean()
std_val = aspect_summary_z["avg_sentiment"].std()

if std_val == 0 or np.isnan(std_val):
    aspect_summary_z["z_score"] = 0.0
else:
    aspect_summary_z["z_score"] = (
        (aspect_summary_z["avg_sentiment"] - mean_val) / std_val
    )

def sentiment_status_z(z):
    if z > 0.5:
        return "🟢 Positif"
    elif z < -0.5:
        return "🔴 Negatif"
    else:
        return "🟡 Netral"

aspect_summary_z["sentiment_status_z"] = aspect_summary_z["z_score"].apply(sentiment_status_z)

z_summary_path = OUTPUT_DIR / "aspect_summary_with_zscore.csv"
aspect_summary_z.to_csv(z_summary_path, encoding="utf-8-sig")

print("\nSUMMARY ASPEK DENGAN Z-SCORE (tanpa 'umum') :")
display(aspect_summary_z)
````

#### Mengapa Z-Score Penting? (Perbandingan Pendekatan)

- **Tanpa Z-Score** (hanya melihat `avg_sentiment` atau `% negative` secara absolut):  
  Hampir semua aspek tampak negatif karena `avg_sentiment` mayoritas bernilai negatif (rata-rata keseluruhan ≈ -0.357).  
  Contoh:  
  - `desain` (-0.81) → terlihat sangat buruk  
  - `baterai` (-0.16) → terlihat “cukup buruk”  
  - `charging` (-0.20) → juga negatif  
  → Pendekatan ini cenderung **over-generalize** dan sulit membedakan mana yang benar-benar **pain point utama** vs hanya ikut-ikutan tren negatif keseluruhan produk.

- **Dengan Z-Score** (normalisasi relatif):  
  Z-score mengukur **seberapa jauh** `avg_sentiment` suatu aspek menyimpang dari **rata-rata semua aspek** (dalam satuan standar deviasi).  
  
  Pertama, kita menghitung rata-rata ($\mu$) dan standar deviasi ($\sigma$) dari seluruh nilai sentimen antar aspek. Kemudian, Z-score untuk masing-masing aspek dihitung dengan rumus:

$$Z = \frac{\text{avg\_sentiment} - \mu}{\sigma}$$

  Hasil:  
  - Nilai positif → sentimen **lebih baik** dari rata-rata aspek lain  
  - Nilai negatif → sentimen **lebih buruk** dari rata-rata  
  - Semakin ekstrem nilai absolutnya → semakin signifikan perbedaannya  

  Kategori status berdasarkan ambang ±0.5 (untuk menghindari sensitivitas berlebih pada fluktuasi kecil):  
  - 🟢 **Positif** → Z-score > +0.5  
  - 🟡 **Netral**  → -0.5 ≤ Z-score ≤ +0.5  
  - 🔴 **Negatif** → Z-score < -0.5  

Pendekatan Z-score membuat analisis jauh lebih **tajam** dan **berorientasi aksi**, karena fokus pada **deviasi relatif** — bukan hanya nilai absolut yang bisa bias karena sentimen keseluruhan produk sedang rendah.

### 📊 Aspect-Based Categorization: Mengapa Menggunakan Z-Score?

Setelah menghitung rata-rata sentimen (`avg_sentiment`) untuk setiap aspek fitur produk, kita dihadapkan pada satu pertanyaan kritis: **Bagaimana cara objektif menentukan apakah suatu aspek benar-benar kuat atau lemah dibandingkan aspek lainnya?** Jika kita hanya menggunakan rata-rata absolut (tanpa Z-score), kita bisa terkecoh. Misalnya, sebuah fitur memiliki skor sentimen **0.15**. Sekilas angka ini terlihat "positif" karena di atas nol. Namun, jika rata-rata sentimen keseluruhan fitur produk ternyata berada di angka **0.40**, maka skor 0.15 sebenarnya **relatif rendah (underperforming)**.

Di sinilah kita menggunakan **Z-score**. Z-score tidak melihat nilai secara absolut, melainkan mengukur posisi relatif (*penyimpangan*) suatu aspek dibandingkan rata-rata ekosistemnya. 

#### 🧮 Logika Matematis & Kategorisasi
Pertama, kita menghitung rata-rata ($\mu$) dan standar deviasi ($\sigma$) dari seluruh nilai sentimen antar aspek. Kemudian, Z-score untuk masing-masing aspek dihitung dengan rumus:

$$Z = \frac{\text{avg\_sentiment} - \mu}{\sigma}$$

Berdasarkan nilai $Z$ yang dihasilkan, kita mengelompokkan persepsi pasar ke dalam tiga ambang batas (±0.5 digunakan agar model tidak terlalu sensitif/reaktif terhadap perbedaan angka yang sangat kecil):
- 🟢 **Positif (Z-score > 0.5):** Performa sentimen jauh di atas rata-rata fitur lainnya.
- 🔴 **Negatif (Z-score < -0.5):** Performa sentimen jauh di bawah rata-rata (tertinggal).
- 🟡 **Netral (Z-score antara -0.5 s/d 0.5):** Performa standar, seimbang dengan keseluruhan produk.

#### 6.5. Prioritization Matrix: Impact vs Sentiment

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
from scipy.stats import zscore

# LOAD FOLDER ASPECT TERBARU
PROJECT_ROOT = Path(".").resolve()
ASPECT_ROOT = PROJECT_ROOT / "dungeon" / "summary_analysis"

latest_aspect_dir = sorted(ASPECT_ROOT.glob("aspect_based_run_*"))[-1]
print("Using aspect folder:", latest_aspect_dir)

# LOAD DATA EXPLODED
exploded_path = latest_aspect_dir / "aspect_tagged_exploded.csv"

if not exploded_path.exists():
    raise FileNotFoundError(f"File exploded tidak ditemukan: {exploded_path}")

df_exploded = pd.read_csv(exploded_path).copy()

required_cols = ["aspects", "clean_text", "polarity"]
missing = [c for c in required_cols if c not in df_exploded.columns]
if missing:
    raise ValueError(
        f"Kolom wajib tidak lengkap. Missing: {missing}\n"
        f"Kolom tersedia: {list(df_exploded.columns)}"
    )

# EXCLUDE 'umum'
df_exploded["aspects"] = df_exploded["aspects"].astype(str).str.strip().str.lower()
df_analysis = df_exploded[df_exploded["aspects"] != "umum"].copy()

if df_analysis.empty:
    raise ValueError("Data analisis kosong setelah menghapus kategori 'umum'.")

# BUILD METRICS
aspect_metrics = (
    df_analysis
    .groupby("aspects", as_index=False)
    .agg(
        frequency=("clean_text", "count"),
        avg_sentiment=("polarity", "mean")
    )
)

MIN_MENTIONS = 30
aspect_metrics = aspect_metrics[aspect_metrics["frequency"] >= MIN_MENTIONS].copy()

if aspect_metrics.empty:
    raise ValueError(
        f"Tidak ada aspek dengan minimum mention >= {MIN_MENTIONS}. "
        "Turunkan nilai MIN_MENTIONS jika perlu."
    )

# Z-SCORE
std_val = aspect_metrics["avg_sentiment"].std()

if std_val == 0 or np.isnan(std_val):
    aspect_metrics["sentiment_z"] = 0.0
else:
    aspect_metrics["sentiment_z"] = zscore(aspect_metrics["avg_sentiment"])

# COLOR RULE
def color_rule(z):
    if z < -1.0:
        return "#c0392b"   # merah kuat
    elif z < -0.2:
        return "#e67e22"   # orange
    else:
        return "#27ae60"   # hijau

aspect_metrics["color"] = aspect_metrics["sentiment_z"].apply(color_rule)

# BUBBLE SIZE SCALING
# Supaya bubble tidak terlalu kecil / terlalu besar
min_size = 120
max_size = 700

freq_min = aspect_metrics["frequency"].min()
freq_max = aspect_metrics["frequency"].max()

if freq_min == freq_max:
    aspect_metrics["bubble_size"] = 300
else:
    aspect_metrics["bubble_size"] = min_size + (
        (aspect_metrics["frequency"] - freq_min) / (freq_max - freq_min)
    ) * (max_size - min_size)

# THRESHOLDS
x_threshold = 0.0
freq_median = aspect_metrics["frequency"].median()

# PLOT
fig, ax = plt.subplots(figsize=(12, 8))

scatter = ax.scatter(
    aspect_metrics["sentiment_z"],
    aspect_metrics["frequency"],
    s=aspect_metrics["bubble_size"],
    c=aspect_metrics["color"],
    edgecolor="black",
    linewidth=1.2,
    alpha=0.9
)

# LABEL ASPEK
for _, row in aspect_metrics.iterrows():
    ax.text(
        row["sentiment_z"] + 0.03,
        row["frequency"],
        row["aspects"],
        fontsize=10,
        weight="bold",
        va="center"
    )

# GARIS KUADRAN
ax.axvline(x_threshold, color="black", linestyle="--", alpha=0.6, linewidth=1.5)
ax.axhline(freq_median, color="black", linestyle="--", alpha=0.6, linewidth=1.5)

# FINAL TOUCH AXIS
ax.set_title("Prioritization Matrix: Impact vs Sentiment (Final SetFit)", fontsize=18)
ax.set_xlabel("Sentiment Score (Z-score)", fontsize=13)
ax.set_ylabel("Frequency (Volume Pembicaraan)", fontsize=13)
ax.grid(alpha=0.25)

# kasih margin kiri-kanan biar label ga mepet
x_min = aspect_metrics["sentiment_z"].min()
x_max = aspect_metrics["sentiment_z"].max()
x_pad = max(0.15, (x_max - x_min) * 0.15)

ax.set_xlim(x_min - x_pad, x_max + x_pad)

y_max = aspect_metrics["frequency"].max()
ax.set_ylim(0, y_max * 1.12)

# LEGEND QUADRANT DI LUAR PLOT
fig.text(
    0.03, 0.89,
    "CRITICAL ISSUES\nHigh Volume • Negative",
    color="red",
    fontsize=11,
    bbox=dict(facecolor="white", edgecolor="red", boxstyle="round,pad=0.4")
)

fig.text(
    0.83, 0.89,
    "SELLING POINTS\nHigh Volume • Positive",
    color="green",
    fontsize=11,
    bbox=dict(facecolor="white", edgecolor="green", boxstyle="round,pad=0.4")
)

fig.text(
    0.03, 0.10,
    "MINOR ISSUES\nLow Volume • Negative",
    color="darkred",
    fontsize=10,
    bbox=dict(facecolor="white", edgecolor="gray", boxstyle="round,pad=0.4")
)

fig.text(
    0.83, 0.10,
    "NICE TO HAVE\nLow Volume • Positive",
    color="darkgreen",
    fontsize=10,
    bbox=dict(facecolor="white", edgecolor="gray", boxstyle="round,pad=0.4")
)

# OPTIONAL: NOTE KECIL
fig.text(
    0.5, 0.02,
    f"Min mentions filter = {MIN_MENTIONS} | Median frequency = {freq_median:.0f} | Vertical split = Z-score {x_threshold:.1f}",
    ha="center",
    fontsize=9,
    color="dimgray"
)

plt.tight_layout(rect=[0.02, 0.05, 0.98, 0.97])

save_path = latest_aspect_dir / "chart_matrix_priority_advanced.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()

print("Saved:", save_path)
````

Matrix ini merangkum seluruh analisis dalam satu peta yang mudah dipahami, sumbu horizontal menunjukkan sentimen (kiri = lebih negatif, kanan = lebih positif), dan sumbu vertikal menunjukkan seberapa sering aspek tersebut dibicarakan (volume).

Fokus utama ada di kiri atas — area High Volume & Negative. Di sini terdapat `kamera` dan `desain`. Artinya, dua fitur ini bukan hanya banyak dikeluhkan, tetapi juga sering dibahas. Dari sudut pandang bisnis, ini adalah sinyal risiko yang nyata karena dapat memengaruhi persepsi pasar dan keputusan pembelian.

Di sisi kanan bawah terdapat `baterai`, `performa`, `software`, dan `layar`. Persepsinya relatif positif, namun volumenya tidak sebesar `kamera`. Ini menunjukkan fitur-fitur tersebut cukup solid, tetapi belum menjadi keunggulan dominan yang mendorong percakapan publik.

Secara sederhana, matrix ini membantu menjawab dua pertanyaan penting untuk pengambilan keputusan:
- Mana yang harus segera diperbaiki? ->  `Kamera` dan `desain`.
- Mana yang bisa dipertahankan dan diperkuat dalam komunikasi? -> `Baterai` dan `layar`.

Visual ini mengubah data teknis menjadi arah strategi yang jelas dan actionable.

#### 6.6. Cross-Channel Sentiment Heatmap (Konsistensi Persepsi Antar Reviewer)

````python
import matplotlib.pyplot as plt
import seaborn as sns

if "source_channel" not in df_exploded.columns:
    raise ValueError("Kolom `source_channel` tidak ditemukan di dataset.")

MIN_MENTIONS = 30

aspect_counts = (
    df_exploded
    .groupby("aspects")
    .size()
)

valid_aspects = aspect_counts[aspect_counts >= MIN_MENTIONS].index

df_heat = df_exploded[df_exploded["aspects"].isin(valid_aspects)].copy()

heatmap_data = df_heat.pivot_table(
    index="aspects",
    columns="source_channel",
    values="polarity",
    aggfunc="mean"
)

mask = heatmap_data.isna()

plt.figure(figsize=(12, 8))

sns.heatmap(
    heatmap_data,
    mask=mask,
    cmap="RdYlGn",           # warna seperti gambar
    center=0,                # pusat warna = netral
    vmin=-1,                 # skala minimum
    vmax=1,                  # skala maksimum
    linewidths=0.5,
    annot=True,
    fmt=".2f",
    cbar_kws={"label": "Avg Sentiment Score (-1 → +1)"}
)

plt.title(
    "Cross-Channel Sentiment Heatmap\n(Konsistensi Persepsi Antar Reviewer)",
    fontsize=14
)

plt.ylabel("Aspek Produk")
plt.xlabel("Channel YouTube")

plt.tight_layout()

save_path = latest_aspect_dir / "chart_heatmap_consistency.png"

plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("Saved:", save_path)
````

Heatmap ini menunjukkan bagaimana persepsi terhadap setiap aspek berbeda atau konsisten di antara tiga channel reviewer: `GadgetIn`, `Jagat Review`, dan `Youtuber Cupu`. Skala warna menggambarkan rata-rata sentimen dari -1 (merah/negatif) hingga +1 (hijau/positif).

- Hal paling menonjol adalah `desain`, yang konsisten negatif di semua channel. Nilainya berada di kisaran -0.74 hingga -0.83. Ini menunjukkan bahwa persepsi negatif terhadap `desain` bukan bias satu reviewer saja, melainkan pola yang cukup seragam di berbagai komunitas audiens.

- Untuk `kamera`, ketiganya juga cenderung negatif, meskipun dengan intensitas berbeda. GadgetIn dan Youtuber Cupu lebih negatif dibanding Jagat Review. Artinya, ada sedikit variasi framing antar reviewer, tetapi arah persepsinya tetap sama.

- Pada `harga` dan `software`, sentimen cenderung negatif ringan hingga moderat, terutama di Jagat Review. Ini menunjukkan kemungkinan perbedaan sudut pandang editorial atau audiens.

- Menariknya, `baterai` terlihat cukup berbeda antar channel. Jagat Review menunjukkan sentimen positif, sementara dua channel lainnya cenderung netral hingga sedikit negatif. Ini menunjukkan adanya variasi pengalaman atau ekspektasi audiens.

Secara keseluruhan, heatmap ini membantu menjawab pertanyaan penting untuk stakeholder:
Apakah persepsi negatif ini hanya efek framing satu reviewer, atau konsisten lintas channel?

Dalam kasus `desain` dan `kamera`, jawabannya jelas, konsisten lintas channel. Ini memperkuat bahwa isu tersebut bersifat struktural, bukan sekadar opini subjektif satu komunitas.

### **7. LLM Report - Gemini API**

Setelah seluruh tahapan analisis kuantitatif selesai, mulai dari `fine-tuned sentiment classification`, `aspect-based categorization`, `Z-score normalization`, `pain point prioritization`, hingga `cross-channel comparison`, tahap berikutnya adalah mengubah temuan analitis menjadi rekomendasi strategis yang siap dikonsumsi oleh eksekutif.

Disini saya mencoba pendekatan untuk menintegrasikan LLM (Gemini API), yang mana pendekatan ini tidak menggantikan model sentimen yang telah dibangun, melainkan berfungsi sebagai lapisan interpretatif tingkat lanjut. Seluruh output numerik dari pipeline sebelumnya dirangkum menjadi struktur laporan strategis yang terarah, ringkas, dan berorientasi pada keputusan.

Secara alur:
1. Data Summary (KPI inti) dari output final SetFit. Menghitung total komentar, distribusi positif/negatif/netral, lalu membuat Net Sentiment (pos% - neg%). Ini jadi “headline metrics” yang cepat dipahami eksekutif.
2. Aspect Insights (bahan narasi) dari aspect_summary. 
3. Prompt engineering untuk format laporan konsultatif, Prompt menempatkan model sebagai “Senior Strategy Consultant” dan mengunci struktur output, Executive Summary -> Diagnosis -> Risk -> 90-day plan -> implikasi product/engineering -> strategi marketing -> outlook. Ini membuat output Gemini bukan sekadar rangkuman, tapi draft report yang siap dipresentasikan.

>Secara konseptual, Analisis tidak berhenti pada visualisasi dan tabel, tetapi berlanjut menjadi rekomendasi yang dapat digunakan untuk pengambilan keputusan. Ini adalah tahap di mana pipeline machine learning bertransformasi menjadi alat intelligence yang mendukung strategi produk dan positioning pasar.


````python
# Libraries
from pathlib import Path
from datetime import datetime
import json
import os

import pandas as pd
from dotenv import load_dotenv
from google import genai
````

````python
client = genai.Client()

models = list(client.models.list())

for m in models[:30]:
    print(m.name)
````

````python
# 0) CONFIG
load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")
if not GEMINI_API_KEY:
    raise EnvironmentError("GEMINI_API_KEY belum diset di file .env")

PROJECT_ROOT = Path(".").resolve()

# ambil hasil aspect run terbaru
ASPECT_ROOT = PROJECT_ROOT / "dungeon" / "summary_analysis"
latest_aspect_dir = sorted(ASPECT_ROOT.glob("aspect_based_run_*"))[-1]

INPUT_DATA_PATH = PROJECT_ROOT / "dungeon" / "processed_dataset" / "sentiment_analyzed_setfit_final.csv"
ASPECT_Z_PATH = latest_aspect_dir / "aspect_summary_with_zscore.csv"

RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
OUTPUT_DIR = PROJECT_ROOT / "dungeon" / "summary_analysis" / f"llm_report_run_{RUN_ID}"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

REPORT_TXT_PATH = OUTPUT_DIR / "final_recommendation.txt"
PROMPT_PATH = OUTPUT_DIR / "prompt_used.txt"
META_PATH = OUTPUT_DIR / "run_meta.json"

MODEL_NAME = "gemini-3.1-pro-preview"
TARGET_PRODUCT = "Xiaomi Note 15 5G"
MIN_MENTIONS = 30

print("Project root      :", PROJECT_ROOT)
print("Input dataset     :", INPUT_DATA_PATH)
print("Aspect Z-score    :", ASPECT_Z_PATH)
print("Output dir        :", OUTPUT_DIR)
print("Model             :", MODEL_NAME)
````

````python
# 1) LOAD DATA
if not INPUT_DATA_PATH.exists():
    raise FileNotFoundError(f"File input tidak ditemukan: {INPUT_DATA_PATH}")

if not ASPECT_Z_PATH.exists():
    raise FileNotFoundError(f"File aspect z-score tidak ditemukan: {ASPECT_Z_PATH}")

df = pd.read_csv(INPUT_DATA_PATH).copy()
aspect_summary = pd.read_csv(ASPECT_Z_PATH, index_col=0).copy()

LABEL_COL_FINAL = "sentiment_label_final"

if LABEL_COL_FINAL not in df.columns:
    raise ValueError(
        f"Kolom '{LABEL_COL_FINAL}' tidak ditemukan. Kolom tersedia: {list(df.columns)}"
    )

print(f"\nLoaded final sentiment dataset: {len(df):,} komentar")
print(f"Loaded aspect z-score summary : {len(aspect_summary):,} aspek")
````

````python
# 2) DATA SUMMARY & STATISTICS
df[LABEL_COL_FINAL] = df[LABEL_COL_FINAL].astype(str).str.lower().str.strip()

total_com = len(df)

sentiment_counts = df[LABEL_COL_FINAL].value_counts(normalize=False)
sentiment_pct = df[LABEL_COL_FINAL].value_counts(normalize=True) * 100

pos_count = int(sentiment_counts.get("positive", 0))
neg_count = int(sentiment_counts.get("negative", 0))
neu_count = int(sentiment_counts.get("neutral", 0))

pos_pct = float(sentiment_pct.get("positive", 0))
neg_pct = float(sentiment_pct.get("negative", 0))
neu_pct = float(sentiment_pct.get("neutral", 0))

net_sentiment = pos_pct - neg_pct

# 3) ASPECT INSIGHTS (Z-SCORE ONLY)
try:
    aspect_focus = aspect_summary[aspect_summary["total_mentions"] >= MIN_MENTIONS].copy()

    # top negative by z-score
    top_negative_aspects = (
        aspect_focus
        .sort_values(["z_score", "total_mentions"], ascending=[True, False])
        .head(3)
    )

    # top positive by z-score
    top_positive_aspects = (
        aspect_focus
        .sort_values(["z_score", "total_mentions"], ascending=[False, False])
        .head(3)
    )

    top_complaints = "\n".join([
        f"- {idx}: z_score={row['z_score']:.2f}, mentions={int(row['total_mentions'])}, status={row.get('sentiment_status_z', '-')}"
        for idx, row in top_negative_aspects.iterrows()
    ])

    top_praises = "\n".join([
        f"- {idx}: z_score={row['z_score']:.2f}, mentions={int(row['total_mentions'])}, status={row.get('sentiment_status_z', '-')}"
        for idx, row in top_positive_aspects.iterrows()
    ])

except Exception:
    top_complaints = "- Belum tersedia"
    top_praises = "- Belum tersedia"

today_str = datetime.now().strftime("%d %B %Y")
````

````python
# 4) BUILD PROMPT
prompt = f"""
Tanggal: {today_str}

Anda berperan sebagai Senior Strategy Consultant yang ditugaskan oleh
Xiaomi Global Market Intelligence Team untuk menyusun laporan strategis
berbasis data sentimen pasar.

Produk utama yang dianalisis:
{TARGET_PRODUCT}

Penting:
- Fokus analisis HARUS tetap pada Xiaomi Note 15 5G / Redmi Note 15 series
- Jika ada komentar yang memuji kompetitor, perlakukan itu sebagai tekanan kompetitif terhadap Xiaomi
- Gunakan data berikut sebagai dasar utama
- Gunakan penalaran strategis, bukan sekadar deskripsi ulang angka

========================================================
DATA INTELIJEN PASAR (YouTube Sentiment Analysis)
========================================================

Total Sampel: {total_com:,} komentar

Brand Health Index (Net Sentiment):
{net_sentiment:.1f} / 100

Distribusi Sentimen:
- Positif : {pos_pct:.1f}% ({pos_count:,})
- Negatif : {neg_pct:.1f}% ({neg_count:,})
- Netral  : {neu_pct:.1f}% ({neu_count:,})

Top Performing Features (berdasarkan Z-score, minimal {MIN_MENTIONS} mention):
{top_praises}

Top Pain Points (berdasarkan Z-score, minimal {MIN_MENTIONS} mention):
{top_complaints}

Catatan interpretasi:
- Z-score positif berarti aspek tersebut relatif lebih kuat dibanding rata-rata aspek lain
- Z-score negatif berarti aspek tersebut relatif lebih lemah dibanding rata-rata aspek lain
- Gunakan Z-score untuk prioritas relatif antar aspek, bukan sebagai skor absolut kepuasan

========================================================
DELIVERABLE YANG DIHARAPKAN
========================================================

Susun laporan dengan struktur berikut:

1. EXECUTIVE SUMMARY
- Tentukan status peluncuran: STRONG / STABLE / AT RISK
- Jelaskan kondisi pasar saat ini secara objektif
- Berikan 3–5 key takeaways yang paling penting secara bisnis

2. MARKET DIAGNOSIS
- Apa kekuatan kompetitif utama Xiaomi Note 15?
- Apa titik friksi terbesar user?
- Apakah problem yang muncul bersifat sistemik, lokal, atau minor?

3. RISK ASSESSMENT
- Jika pain points tidak ditangani, apa implikasi 3–6 bulan ke depan?
- Risiko terhadap brand perception?
- Risiko terhadap repeat purchase dan loyalty?

4. STRATEGIC PRIORITIES (90-Day Plan)
Kelompokkan menjadi:
- Immediate Fix (High Impact, Low Effort)
- Structural Improvement (High Impact, High Effort)
- Monitor Only (Low Impact)

5. PRODUCT & ENGINEERING IMPLICATION
- Rekomendasi hardware adjustment bila relevan
- Rekomendasi software / HyperOS / MIUI optimization
- Trade-off analysis jika perlu

6. MARKETING & POSITIONING STRATEGY
- Fitur mana yang perlu di-double down?
- Narasi mana yang perlu dikontrol?
- Apakah perlu counter-positioning terhadap persepsi negatif?

7. FORWARD OUTLOOK
- Apakah positioning produk ini sustainable?
- Apakah produk ini punya defensible advantage?
- Apakah perlu repositioning di segmen harga?

========================================================
GAYA PENULISAN
========================================================

- Gunakan bahasa strategic namun natural
- Hindari kalimat generik seperti "perlu ditingkatkan"
- Setiap insight harus punya implikasi bisnis
- Rekomendasi harus actionable dan realistis
- Gunakan bullet points jika membantu
- Hindari terlalu teknis kecuali relevan untuk keputusan
- Tulis dalam Bahasa Indonesia
- Buat ringkas tetapi kuat, seperti memo untuk executive leadership team

Buat laporan yang tajam, relevan, dan berbasis insight.
""".strip()

# simpan prompt
with open(PROMPT_PATH, "w", encoding="utf-8") as f:
    f.write(prompt)
````

````python
# 5) GENERATE REPORT
client = genai.Client(api_key=GEMINI_API_KEY)

print(f"\nMenggunakan model: {MODEL_NAME}")

response = client.models.generate_content(
    model=MODEL_NAME,
    contents=prompt
)

report_text = response.text

print("\n" + "=" * 70)
print("LAPORAN STRATEGIS SENTIMENT ANALYSIS (XIAOMI INTERNAL MEMO)")
print("=" * 70)
print(report_text)
````

````python
# 6) SAVE OUTPUT
# =========================================================
with open(REPORT_TXT_PATH, "w", encoding="utf-8") as f:
    f.write(report_text)

meta = {
    "run_id": RUN_ID,
    "timestamp": datetime.now().isoformat(),
    "model_name": MODEL_NAME,
    "target_product": TARGET_PRODUCT,
    "input_dataset": str(INPUT_DATA_PATH),
    "aspect_zscore_input": str(ASPECT_Z_PATH),
    "output_report": str(REPORT_TXT_PATH),
    "prompt_path": str(PROMPT_PATH),
    "notes": "LLM report generation using Gemini API, based on final SetFit sentiment distribution and aspect z-score summary."
}

with open(META_PATH, "w", encoding="utf-8") as f:
    json.dump(meta, f, ensure_ascii=False, indent=2)

print(f"\nLaporan tersimpan di: {REPORT_TXT_PATH}")
print(f"Prompt tersimpan di : {PROMPT_PATH}")
print(f"Meta tersimpan di   : {META_PATH}")
````

````python
# Libraries for PDF generation
from pathlib import Path
import re
from html import escape

from reportlab.lib.pagesizes import A4
from reportlab.lib.colors import HexColor
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import cm
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    PageBreak,
    ListFlowable,
    ListItem,
)
````

````python
# 0) CONFIG
PROJECT_ROOT = Path(".").resolve()

LLM_ROOT = PROJECT_ROOT / "dungeon" / "summary_analysis"
latest_llm_dir = sorted(LLM_ROOT.glob("llm_report_run_*"))[-1]

INPUT_TXT = latest_llm_dir / "final_recommendation.txt"
OUTPUT_PDF = latest_llm_dir / "Xiaomi_Note_15_Post_Launch_Report.pdf"

print("Project root :", PROJECT_ROOT)
print("LLM run dir  :", latest_llm_dir)
print("Input TXT    :", INPUT_TXT)
print("Output PDF   :", OUTPUT_PDF)
````

````python
# 1) LOAD CONTENT
# =========================================================
if not INPUT_TXT.exists():
    raise FileNotFoundError(f"File tidak ditemukan: {INPUT_TXT}")

raw_text = INPUT_TXT.read_text(encoding="utf-8").strip()
lines = [line.rstrip() for line in raw_text.split("\n")]

print(f"\nLoaded TXT report: {len(lines):,} lines")

# =========================================================
# 2) PDF SETUP
# =========================================================
doc = SimpleDocTemplate(
    str(OUTPUT_PDF),
    pagesize=A4,
    leftMargin=2.8 * cm,
    rightMargin=2.8 * cm,
    topMargin=3.5 * cm,
    bottomMargin=2.5 * cm,
)

styles = getSampleStyleSheet()

def add_style(name, **kwargs):
    if name not in styles:
        styles.add(ParagraphStyle(name=name, **kwargs))

# cover styles
add_style(
    "CoverMainTitle",
    fontSize=26,
    leading=30,
    alignment=TA_CENTER,
    spaceAfter=12,
    textColor=HexColor("#0D47A1"),
    fontName="Helvetica-Bold",
)

add_style(
    "CoverProduct",
    fontSize=20,
    leading=24,
    alignment=TA_CENTER,
    spaceAfter=0,
    textColor=HexColor("#1565C0"),
    fontName="Helvetica",
)

# content styles
add_style(
    "Section",
    fontSize=15.5,
    leading=19,
    spaceBefore=16,
    spaceAfter=7,
    textColor=HexColor("#1565C0"),
    fontName="Helvetica-Bold",
)

add_style(
    "Subsection",
    fontSize=13,
    leading=16,
    spaceBefore=12,
    spaceAfter=6,
    leftIndent=8,
    textColor=HexColor("#424242"),
    fontName="Helvetica-Bold",
)

add_style(
    "Body",
    fontSize=10.7,
    leading=14.5,
    alignment=TA_JUSTIFY,
    spaceAfter=7,
    spaceBefore=3,
)

add_style(
    "Bullet",
    fontSize=10.5,
    leading=14,
    leftIndent=26,
    bulletIndent=13,
    spaceAfter=5,
)

story = []

# 3) MARKDOWN -> REPORTLAB CONVERTER
def md_to_rl(text: str) -> str:
    if not isinstance(text, str):
        return ""
    text = escape(text)
    text = re.sub(r"\*\*(.+?)\*\*", r"<b>\1</b>", text)
    text = text.replace("*", "")
    return text

# 4) COVER PAGE
story.append(Spacer(1, 8 * cm))
story.append(Paragraph("Post-Launch Market Intelligence Report", styles["CoverMainTitle"]))
story.append(Spacer(1, 1.2 * cm))
story.append(Paragraph("Xiaomi Note 15 5G", styles["CoverProduct"]))
story.append(Spacer(1, 10 * cm))
story.append(PageBreak())

# 5) CONTENT PARSER
bullet_buffer = []
prev_was_section = False

def flush_bullets():
    if not bullet_buffer:
        return

    items = [
        ListItem(Paragraph(md_to_rl(txt.strip()), styles["Bullet"]))
        for txt in bullet_buffer
    ]

    story.append(
        ListFlowable(
            items,
            bulletType="bullet",
            bulletFontName="Helvetica",
            bulletFontSize=10.8,
            leftPadding=6,
            spaceBetween=3.5,
        )
    )
    bullet_buffer.clear()

for line in lines:
    stripped = line.strip()

    # skip separator / kosong
    if re.match(r"^[-—━•*_=]{3,}$", stripped) or not stripped:
        flush_bullets()
        story.append(Spacer(1, 5))
        continue

    # section utama markdown: ### 
    if stripped.startswith("### "):
        flush_bullets()

        if prev_was_section:
            story.append(PageBreak())
        else:
            story.append(Spacer(1, 8))

        section_title = stripped[4:].strip()
        story.append(Paragraph(md_to_rl(section_title), styles["Section"]))
        story.append(Spacer(1, 6))
        prev_was_section = True
        continue

    # section bernomor: "1. EXECUTIVE SUMMARY"
    if re.match(r"^\d+\.\s+[A-Z].*", stripped):
        flush_bullets()
        story.append(Spacer(1, 8))
        story.append(Paragraph(md_to_rl(stripped), styles["Section"]))
        story.append(Spacer(1, 6))
        prev_was_section = True
        continue

    # subsection: **...** atau ...
    if (
        (stripped.startswith("**") and stripped.endswith("**"))
        or (stripped.endswith(":") and len(stripped) < 70 and not stripped.isupper())
    ):
        flush_bullets()
        clean = stripped.strip("*").strip(":").strip()
        story.append(Paragraph(f"<b>{md_to_rl(clean)}</b>", styles["Subsection"]))
        story.append(Spacer(1, 5))
        continue

    # bullet line
    if stripped.startswith(("-", "•", "*", "→", "–")) or re.match(r"^\s*[-*•→–]\s", line):
        bullet_text = re.sub(r"^[-*•→–]\s*", "", stripped).strip()
        bullet_buffer.append(bullet_text)
        continue

    # paragraf biasa
    flush_bullets()
    story.append(Paragraph(md_to_rl(stripped), styles["Body"]))
    story.append(Spacer(1, 4))

flush_bullets()

# 6) BUILD PDF
doc.build(story)

print(f"\nPDF selesai -> {OUTPUT_PDF.resolve()}")
print(f"Ukuran file : {OUTPUT_PDF.stat().st_size / 1024:.1f} KB")
````
