### 1.2 - Proses Ekstraksi Komentar dari Youtube

Langkah pertama Untuk melakukan analisis sentimen adalah memperoleh data teks berupa komentar dari video YouTube terkait. YouTube tidak menyediakan fitur ekspor komentar secara langsung, sehingga kita perlu melakukan web scraping menggunakan [YouTube API](https://developers.google.com/youtube/v3) yang resmi dari Google.

```python
def extract_youtube_comments(video_id, api_key, max_comments=7000):
    """
    Mengambil komentar dari YouTube dengan paginasi dan error handling.
    """
    youtube = build('youtube', 'v3', developerKey=api_key)
    comments = []
    next_page_token = None
    fetched = 0
    
    # Progress bar untuk estimasi
    pbar = tqdm(total=max_comments, desc=f"Scraping {video_id}", leave=False)

    try:
        while fetched < max_comments:
            request = youtube.commentThreads().list(
                part='snippet',
                videoId=video_id,
                maxResults=100,
                pageToken=next_page_token,
                textFormat='plainText',
                order="time" # Opsi "relevance" untuk mengambil komentar terpopuler/relevan
            )
            response = request.execute()

            for item in response.get('items', []):
                snippet = item['snippet']['topLevelComment']['snippet']
                comments.append({
                    "video_id": video_id,
                    "author": snippet.get('authorDisplayName'),
                    "text": snippet.get('textDisplay', ''),
                    "like_count": snippet.get('likeCount', 0),
                    "published_at": snippet.get('publishedAt')
                })
                fetched += 1
                pbar.update(1)
                
                if fetched >= max_comments:
                    break

            next_page_token = response.get('nextPageToken')
            if not next_page_token:
                break
                
    except Exception as e:
        print(f"Error pada video {video_id}: {e}")
    finally:
        pbar.close()

    return pd.DataFrame(comments)
```

```python
all_dfs = []
print("Memulai Ekstraksi Data Multi-Channel")

for video in VIDEO_LIST:
    print(f"-> Processing: {video['channel']}", end=" ")  # cetak tanpa enter
    
    df_vid = extract_youtube_comments(video["id"], YOUTUBE_API_KEY, max_comments=7000)
    
    if not df_vid.empty:
        df_vid["source_channel"] = video["channel"]
        df_vid["video_title"] = video["title"]
        all_dfs.append(df_vid)
        
        print(f"(Berhasil scrap {len(df_vid):,} komentar)")
        
        # Menyimpan raw per video
        filename = f"{OUTPUT_DIR}/raw_{video['channel'].replace(' ', '_')}.csv"
        df_vid.to_csv(filename, index=False)
    else:
        print("(Gagal / tidak ada komentar)")

# Menggabungkan semua data
if all_dfs:
    df_raw = pd.concat(all_dfs, ignore_index=True)
    print(f"\nTOTAL DATA TERKUMPUL (semua channel): {len(df_raw):,} komentar")
    df_raw.to_csv(f"{OUTPUT_DIR}/Raw Dataset/combined_raw_data.csv", index=False)
else:
    print("\nGagal mengambil data sama sekali.")
```

## **2. Text Preprocessing**

Tahap text preprocessing dilakukan untuk mengubah komentar mentah (raw text) menjadi format yang lebih bersih, konsisten, dan siap dianalisis oleh model NLP.

Komentar YouTube umumnya mengandung bahasa informal, emoji, singkatan, serta karakter yang tidak terstruktur. Tanpa pembersihan, hal ini dapat menurunkan akurasi model dan menyebabkan interpretasi yang tidak konsisten.

Proses preprocessing dilakukan secara sistematis melalui beberapa langkah berikut:
1. `Slang Dictionary`, Mengubah bahasa informal menjadi bentuk baku (misal: “gk” → “tidak”, “renot” → “redmi note”). Langkah ini menjaga konsistensi makna dan membantu model memahami konteks secara lebih akurat.
2. `Demoji`, Mendeteksi dan mengekstrak emoji agar dapat diproses secara terstruktur. Emoji tidak langsung dihapus, tetapi dinormalisasi agar tidak dianggap sebagai token acak.
3. `Menghapus URL & Mention`, Menghilangkan tautan dan mention yang tidak berkontribusi terhadap sentimen atau opini produk.
4. `Normalisasi Huruf Berulang`, Mengubah kata seperti “baguuuus” menjadi “bagus” agar tidak dianggap sebagai token berbeda oleh model.
5. `Replace Phrase Slang`, Menstandarkan frasa slang multi-kata agar tidak terpecah saat proses tokenisasi.
6. `Menghapus Simbol Non-Relevan`, Menghilangkan karakter non-alfabet yang bersifat noise dan tidak memiliki makna semantik.
7. `Token Slang Handling`, Memastikan kata slang dikenali sebagai token yang benar sebelum masuk ke tahap modeling.
8. `Hapus Spasi Ganda`, Membersihkan teks agar konsisten dan menghindari error saat tokenisasi.
9. `Filter Komentar Pendek`,Menghapus komentar sangat pendek (kurang dari 4 kata) karena umumnya tidak memberikan opini bermakna untuk analisis sentimen.

Text preprocessing adalah tahapan yang sangat penting karena dapat meningkatkan:
- Konsistensi bahasa
- Kualitas representasi teks
- Stabilitas tokenisasi
- Akurasi model sentimen

>Kualitas output model sangat bergantung pada kualitas input teks. Oleh karena itu, preprocessing menjadi fondasi penting sebelum masuk ke tahap analisis sentimen dan fine-tuning.

```python
# Pre-compile regex biar lebih cepat + rapi
RE_URL_MENTION = re.compile(r"http\S+|www\S+|@\w+")
RE_REPEAT = re.compile(r"(.)\1{2,}")  # >= 3 repetisi
RE_NON_TEXT = re.compile(r"[^a-z0-9\s]")  # Keep angka, agar "5g", "120hz" tidak hilang
RE_MULTI_SPACE = re.compile(r"\s+")

# Menangani phrase slang terlebih dahulu (multi-word), baru token-level
# contoh: "gak bisa", "worth it"
PHRASE_SLANG = {k: v for k, v in slang_dict.items() if " " in k}
WORD_SLANG = {k: v for k, v in slang_dict.items() if " " not in k}

# Build regex phrase dengan boundary agar tidak nyangkut substring aneh
# Mengurutkan phrase terpanjang terlebih dahulu biar gak ketimpa
phrase_patterns = sorted(PHRASE_SLANG.keys(), key=len, reverse=True)
RE_PHRASES = [
    (re.compile(rf"\b{re.escape(p)}\b"), PHRASE_SLANG[p])
    for p in phrase_patterns
]

def clean_text_advanced(text, keep_numeric=True, max_repeat=2):
    if not isinstance(text, str) or not text.strip():
        return ""

    text = text.lower().strip()

    # 2) Mendeteksi dan mengekstrak emoji dari teks agar bisa diproses secara terstruktur.
    text = demoji.replace(text, " ")

    # 3) Hapus URL & Mention
    text = RE_URL_MENTION.sub(" ", text)

    # 4) Normalisasi huruf berulang: "baaaagus" -> "baagus" (maks 2)
    def _reduce(m):
        ch = m.group(1)
        return ch * max_repeat
    text = RE_REPEAT.sub(_reduce, text)

    # 5) Replace phrase slang terlebih dahulu sebelum simbol dibuang
    # biar "gak bisa" / "worth it" terpakai
    for pattern, repl in RE_PHRASES:
        text = pattern.sub(repl, text)

    # 6) Hapus simbol
    if keep_numeric:
        text = RE_NON_TEXT.sub(" ", text)
    else:
        text = re.sub(r"[^a-z\s]", " ", text)

    # 7) Token slang (single word)
    words = text.split()
    fixed_words = [WORD_SLANG.get(w, w) for w in words]
    text = " ".join(fixed_words)

    # 8) Hapus spasi ganda
    return RE_MULTI_SPACE.sub(" ", text).strip()
```

Meskipun IndoBERT mampu memahami konteks kalimat secara menyeluruh, terdapat beberapa kondisi di mana model berbasis konteks masih bisa melewatkan sinyal penting, terutama pada domain spesifik seperti gadget review.

Alasan utama:
1. Menangkap Sinyal Domain-Spesifik, Meskipun model memahami konteks, penguatan keyword membantu memastikan bahwa istilah dengan dampak signifikan tidak tereduksi oleh konteks netral di sekitarnya. Beberapa kata dalam dunia gadget memiliki bobot opini yang kuat, misalnya:

4. Membuat Baseline Lebih Stabil Sebelum Fine-Tuning, Hybrid approach (context + keyword signal) menghasilkan baseline yang lebih kuat, sehingga tahap fine-tuning selanjutnya dapat fokus pada pola bahasa yang lebih kompleks, bukan hanya memperbaiki sinyal eksplisit.

```python
# Optional: compile regex biar lebih cepat
NEG_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["negative"]]
POS_PATTERNS = [re.compile(rf"\b{re.escape(w)}\b") for w in BOOST_WORDS["positive"]]

def _has_match(patterns, text):
    return any(p.search(text) for p in patterns)

def _normalize_label(label: str) -> str:
    label = label.lower()
    if label in {"positive", "neutral", "negative"}:
        return label
    # fallback mapping (kalau ternyata model keluarnya LABEL_X)
    mapping = {"label_0": "negative", "label_1": "neutral", "label_2": "positive"}
    return mapping.get(label, "neutral")

def analyze_hybrid_sentiment(text: str, override_threshold: float = 0.90):
    # Safety
    if not isinstance(text, str) or not text.strip():
        return "neutral", 0.0

    t = text.lower()

    # A. Prediksi Model (token-level truncation)
    try:
        res = sentiment_pipeline(t, truncation=True, max_length=512)[0]
        label = _normalize_label(res.get("label", "neutral"))
        score = float(res.get("score", 0.0))
    except Exception:
        return "neutral", 0.0

    # B. Keyword Boosting (rule-based guardrail)
    has_neg = _has_match(NEG_PATTERNS, t)
    has_pos = _has_match(POS_PATTERNS, t)

    # Kalau ada sinyal negatif kuat, biasanya lebih “tegas” untuk override
    if has_neg:
        if (label != "negative") or (score < override_threshold):
            return "negative", 0.95

    # Positive override hanya jika tidak ada sinyal negatif yang juga kuat
    if has_pos and not has_neg:
        if (label != "positive") or (score < override_threshold):
            return "positive", 0.95

    return label, score
```

```python
# 0) CONFIG
SEED = 42                 # reproducibility sampling
N_PER_GROUP = 75          # sampel per kombinasi source_channel + sentiment_label
VALID_LABELS = ["negative", "neutral", "positive"]

# Single source of truth for paths
PROJECT_ROOT = Path(".").resolve()

ANALYZED_CSV_PATH = PROJECT_ROOT / "Output Analysis" / "Sentiment Analysis Dataset" / "Sentiment_Analyzed.csv"
LABELING_OUT_PATH = PROJECT_ROOT / "Output Analysis" / "Baseline Dataset" / "baseline_labeling_sample.csv"
LABELED_DONE_PATH = PROJECT_ROOT / "Output Analysis" / "Baseline Dataset" / "baseline_labeling_sample_done.csv"

# Artifacts baseline folder
RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
BASELINE_ARTIFACT_DIR = PROJECT_ROOT / "artifacts" / "baseline_hybrid" / f"run_{RUN_ID}"
BASELINE_ARTIFACT_DIR.mkdir(parents=True, exist_ok=True)

# Memastikan folder output labeling ada
LABELING_OUT_PATH.parent.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("Artifacts dir:", BASELINE_ARTIFACT_DIR)

# 1) LOAD DATASET HASIL BASELINE (IndoBERT + keyword boosting)
if not ANALYZED_CSV_PATH.exists():
    raise FileNotFoundError(
        f"File baseline analyzed tidak ditemukan:\n- {ANALYZED_CSV_PATH}\n"
        "Pastikan path benar dan file sudah ada."
    )

df = pd.read_csv(ANALYZED_CSV_PATH)
print(f"\nLoaded baseline analyzed dataset: {len(df)} rows")

# Validasi kolom
required_cols = ["video_id", "author", "text", "clean_text", "source_channel", "sentiment_label", "sentiment_score"]
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise ValueError(
        f"Kolom wajib tidak lengkap. Missing: {missing}\n"
        f"Kolom tersedia: {list(df.columns)}"
    )
```

```python
# 2) BUILD EVAL POOL + STRATIFIED SAMPLING
df_eval_pool = df.copy()

# filter teks kosong
df_eval_pool = df_eval_pool[df_eval_pool["clean_text"].astype(str).str.strip() != ""].copy()

# hanya mengambil label yang valid (jaga-jaga)
df_eval_pool["sentiment_label"] = df_eval_pool["sentiment_label"].astype(str).str.lower().str.strip()
df_eval_pool = df_eval_pool[df_eval_pool["sentiment_label"].isin(VALID_LABELS)].copy()

print(f"Eval pool setelah filter kosong + label valid: {len(df_eval_pool)} rows")

def strat_sample_with_keys(x):
    sc, sl = x.name  # x.name = (source_channel, sentiment_label)
    out = x.sample(min(len(x), N_PER_GROUP), random_state=SEED)
    return out.assign(source_channel=sc, sentiment_label=sl)

df_sample = (
    df_eval_pool
    .groupby(["source_channel", "sentiment_label"], group_keys=False)
    .apply(strat_sample_with_keys, include_groups=False)   # warning hilang, kolom group dibalikin manual
    .reset_index(drop=True)
)

# kolom untuk labeling manual
df_sample["gold_label"] = ""
df_sample["gold_notes"] = ""

# export template labeling
df_sample[[
    "video_id", "author", "text", "clean_text", "source_channel",
    "sentiment_label", "sentiment_score", "gold_label", "gold_notes"
]].to_csv(LABELING_OUT_PATH, index=False)

print("\nLabeling template saved:", LABELING_OUT_PATH)
print("Jumlah sampel untuk labeling:", len(df_sample))
print("\nDistribusi sampel (source_channel x sentiment_label):")
print(df_sample.groupby(["source_channel", "sentiment_label"]).size())

# Menyimpan metadata sampling sebagai artifacts
sampling_meta = {
    "run_id": RUN_ID,
    "seed": SEED,
    "n_per_group": N_PER_GROUP,
    "total_sampled": int(len(df_sample)),
    "groupby": ["source_channel", "sentiment_label"],
    "analyzed_input_csv": str(ANALYZED_CSV_PATH),
    "labeling_template_csv": str(LABELING_OUT_PATH),
}
with open(BASELINE_ARTIFACT_DIR / "sampling_meta.json", "w", encoding="utf-8") as f:
    json.dump(sampling_meta, f, ensure_ascii=False, indent=2)
```

> Untuk sentiment analysis 3 kelas (Negatif / Netral / Positif) berbasis pretrained transformer, sekitar 500–700 (±200 per kelas) labeled samples sudah memadai untuk menghasilkan evaluasi precision, recall, dan F1-score yang stabil dan dengan ukuran sampel ini, perubahan performa model antara baseline dan fine-tuned dapat diinterpretasikan sebagai peningkatan yang signifikan secara praktis, bukan fluktuasi statistik akibat keterbatasan data.

Kenapa memakai ±200 per kelas:
Misalnya :
- TP = 160
- FP = 40 -> Precision = 160/200 = 0.80

Ada 1 eror tambahan:
- TP = 159
- FP = 41 -> Precision = 159/200 = 0.795

Jadi per 1 eror hanya akan turun 0.005% yang menjadikan perubahan kecil menjadi lebih stabil.

## **4.  Fine Tuning (Setfit)**

Untuk meningkatkan performa model tanpa membutuhkan ribuan data berlabel tambahan, digunakan pendekatan SetFit (Sentence Transformer Fine-Tuning) dengan base model:

`sentence-transformers/paraphrase-multilingual-mpnet-base-v2`

Pendekatan ini bekerja pada level kalimat, bukan hanya kata per kata, sehingga mampu menangkap:
- Makna utuh sebuah opini
- Nuansa ekspresi pengguna
- Perbedaan konteks dalam kalimat yang terlihat mirip

Keunggulan utama SetFit adalah kemampuannya meningkatkan akurasi secara signifikan meskipun hanya menggunakan dataset berlabel dalam jumlah terbatas.

Dengan metode ini, model menjadi lebih adaptif terhadap gaya bahasa reviewer gadget Indonesia tanpa memerlukan komputasi besar atau retraining penuh dari awal.

```python
# 0) CONFIG
PROJECT_ROOT = Path(".").resolve()

# Golden dataset (manual labels)
CSV_PATH = PROJECT_ROOT / "Output Analysis" / "Baseline Dataset" / "baseline_labeling_sample_done.csv"

VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42

# Optional: stabilin tokenizer log
os.environ["TOKENIZERS_PARALLELISM"] = "false"

# Artifacts
RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
FINETUNE_DIR = PROJECT_ROOT / "artifacts" / "setfit_finetune" / f"run_{RUN_ID}"
CHECKPOINT_DIR = FINETUNE_DIR / "checkpoints"
FINAL_DIR = FINETUNE_DIR / "final_model"
FINETUNE_DIR.mkdir(parents=True, exist_ok=True)
CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
FINAL_DIR.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("Fine-tune artifacts dir:", FINETUNE_DIR)

# Reproducibility (minimal)
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
```

Sebagai fondasi konfigurasi sebelum proses fine-tuning dimulai. Tujuan utamanya adalah memastikan proses training berjalan rapi, reproducible, dan mudah ditracking lewat artifacts.

Secara garis besar, langkah yang dilakukan adalah:
1. Menentukan Sumber Data Label Manual (Golden Dataset), Pipeline fine-tuning menggunakan baseline labeling sample yang sudah diberi label manual (`baseline_labeling_sample_done.csv`) sebagai golden dataset.
2. Menstandarkan Label & Aturan Validasi, Ditetapkan label yang dianggap valid hanya `negative`, `neutral`, `positive`. Ini memastikan training dan evaluasi tetap konsisten dan menghindari label “liar” akibat typo atau format data yang tidak seragam.
3. Menyiapkan Struktur Artifacts untuk Tracking Training Run, Ini penting untuk workflow yang “semi-production ready”, karena memudahkan audit, rollback, dan reproducibility. Sehingga setiap eksperimen fine-tuning memiliki jejak yang jelas, termasuk:
    - Checkpoints (model intermediate)
	- Final model (model terbaik)
	- Output evaluasi (metrics/report)

```python
# 1) Device auto-detect (GPU -> cuda, else cpu)
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device yang dipakai:", device)

# Optional housekeeping
gc.collect()
if device == "cuda":
    try:
        torch.cuda.empty_cache()
    except Exception:
        pass

# 2) Load Golden Dataset (manual labels)
if not CSV_PATH.exists():
    raise FileNotFoundError(
        f"File '{CSV_PATH}' tidak ditemukan.\n"
        "Pastikan path & nama folder sama persis (case-sensitive di Linux/Workbench)."
    )

df_gold = pd.read_csv(CSV_PATH)
print(f"Data training dimuat (raw): {len(df_gold)} sampel")
```

```python
# 3) Detect kolom text/label -> SetFit butuh: text, label
possible_text_cols  = ["text", "clean_text", "comment", "raw_text"]
possible_label_cols = ["label", "gold_label", "sentiment_label", "sentiment"]

text_col  = next((c for c in possible_text_cols  if c in df_gold.columns), None)
label_col = next((c for c in possible_label_cols if c in df_gold.columns), None)

if text_col is None or label_col is None:
    raise ValueError(
        "Kolom teks/label tidak ketemu.\n"
        f"Kolom tersedia: {list(df_gold.columns)}\n"
        f"Harus ada salah satu teks: {possible_text_cols}\n"
        f"dan salah satu label: {possible_label_cols}"
    )

# IMPORTANT FIX (minimal tapi penting):
df_gold = df_gold[[text_col, label_col]].copy()
df_gold = df_gold.rename(columns={text_col: "text", label_col: "label"})

df_gold["text"]  = df_gold["text"].astype(str).str.strip()
df_gold["label"] = df_gold["label"].astype(str).str.lower().str.strip()

before = len(df_gold)
df_gold = df_gold[(df_gold["text"] != "") & (df_gold["label"].isin(VALID_LABELS))].copy()
after = len(df_gold)

print(f"\nFilter valid -> sebelum: {before} | sesudah: {after} | drop: {before-after}")
print("\nDistribusi label (gold):")
print(df_gold["label"].value_counts(dropna=False))

if len(df_gold) < 20:
    raise ValueError("Data gold terlalu sedikit setelah filtering. Cek isi label/text di CSV.")
```

```python
# 4) Split train/eval (stratified)
# Stratify aman kalau tiap kelas minimal 2
do_stratify = df_gold["label"].nunique() > 1 and df_gold["label"].value_counts().min() >= 2

train_df, eval_df = train_test_split(
    df_gold,
    test_size=0.2,
    random_state=SEED,
    stratify=df_gold["label"] if do_stratify else None
)

# Save snapshot distribusi train/eval (biar auditable)
dist_train = train_df["label"].value_counts()
dist_eval  = eval_df["label"].value_counts()
dist_df = pd.DataFrame({"train": dist_train, "eval": dist_eval}).fillna(0).astype(int)
dist_df.to_csv(FINETUNE_DIR / "label_distribution_train_eval.csv")
print(f"\nTrain: {len(train_df)} | Eval: {len(eval_df)} | Stratify: {do_stratify}")

train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True), preserve_index=False)
eval_dataset  = Dataset.from_pandas(eval_df.reset_index(drop=True), preserve_index=False)
```

Pada tahap fine-tuning, dataset berlabel (golden dataset) dibagi menjadi:
- 80% Training (520 data)
- 20% Evaluation (131 data)

Pembagian ini dilakukan secara stratified, artinya proporsi masing-masing kelas (negative, neutral, positive) dijaga tetap konsisten di train dan eval set.

1. Pembagian 80:20 dipilih karena dapat memberikan data yang cukup untuk model belajar pola bahasa (training set) dan juga Tetap menyisakan data yang cukup untuk evaluasi performa secara objektif (evaluation set). Karena total data hanya 651, penting untuk tetap menjaga keseimbangan antara kemampuan belajar model dan validitas evaluasi.
2. Stratified split memastikan bahwa distribusi label di train dan eval tetap proporsional. Contoh, Jika kelas positive hanya 10% dari total data, maka di train dan eval juga tetap sekitar 10%. Karena project ini memiliki class imbalance (positive lebih sedikit), stratified split menjadi sangat penting. Tanpa stratification, ada risiko:
    - Eval set kekurangan kelas minority
	- Evaluasi menjadi bias
	- F1-score menjadi tidak representatif

```python
# 5) Load Base Model
BASE_MODEL = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"

model = SetFitModel.from_pretrained(
    BASE_MODEL,
    labels=VALID_LABELS
)

# (opsional) pindahin body/head ke device (keep style kamu, dibuat aman)
try:
    model.model_body.to(device)
except Exception:
    pass

if getattr(model, "model_head", None) is not None:
    try:
        model.model_head.to(device)
    except Exception:
        pass
```

```python
# 6) Training config (Workbench-optimized). Kalau OOM otomatis pakai konfigurasi dibawahnya
if device == "cuda":
    NUM_ITERATIONS = 12
    BATCH_SIZE     = 32
    NUM_EPOCHS     = 3
else:
    NUM_ITERATIONS = 6
    BATCH_SIZE     = 16
    NUM_EPOCHS     = 2

args = SetFitTrainingArguments(
    batch_size=BATCH_SIZE,
    num_epochs=NUM_EPOCHS,
    num_iterations=NUM_ITERATIONS,
)

# Evaluasi & saving secukupnya
args.eval_strategy = "epoch"
args.save_strategy = "epoch"
args.load_best_model_at_end = True

# Checkpoint dir
if hasattr(args, "output_dir"):
    args.output_dir = str(CHECKPOINT_DIR)

trainer = SetFitTrainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)
```

Dalam proses fine-tuning SetFit, digunakan konfigurasi berikut:
- NUM_ITERATIONS = 12
- BATCH_SIZE     = 32
- NUM_EPOCHS     = 3

1. `NUM_ITERATIONS = 12`, menentukan berapa kali pasangan kalimat (sentence pairs) dibentuk dan dipelajari dalam contrastive learning. Semakin besar konfigurasinya model makin kuat memahami semantic similarity akan tetapi waktu training meningkat.
2. `BATCH_SIZE = 32`, Batch size adalah jumlah data yang diproses model dalam satu langkah update. Artinya model akan belajar dari 32 kalimat sekaligus sebelum memperbarui bobotnya. Konfigurasi tersebut adalah nilai yang umum dan stabil untuk GPU T4.
    - Batch besar → stabil & cepat (butuh RAM lebih besar)
    - Batch kecil → lebih detail tapi lebih lambat
3. `NUM_EPOCHS = 3`, Epoch adalah jumlah berapa kali seluruh data training dipelajari oleh model. Artinya model melihat seluruh dataset training sebanyak 3 kali. Konfigurasi tersebut cukup untuk adaptasi tanpa membuat model “menghafal” data, karena dataset relatif kecil (±520 data), jadi jika Epoch terlalu banyak → risiko overfitting

```python
# 7) Train
print("\nSedang melatih ulang model (Fine-Tuning SetFit - Round_0)...")
print(f"Config -> device={device}, batch_size={BATCH_SIZE}, num_epochs={NUM_EPOCHS}, num_iterations={NUM_ITERATIONS}")
print(f"Checkpoint dir: {CHECKPOINT_DIR}")
print(f"Final model dir: {FINAL_DIR}")

t0 = time.time()
trainer.train()
t1 = time.time()

train_minutes = (t1 - t0) / 60
print(f"\nTraining selesai. Durasi: {train_minutes:.2f} menit")
```

```python
# 8) Eval on eval_df (explicit) + save metrics
# SetFitTrainer kadang punya evaluate(), tapi paling stabil: predict langsung
y_true = eval_df["label"].tolist()

# predict returns list of labels
y_pred = trainer.model.predict(eval_df["text"].tolist())

acc = accuracy_score(y_true, y_pred)
report_dict = classification_report(y_true, y_pred, digits=4, output_dict=True)

metrics = {
    "run_id": RUN_ID,
    "device": device,
    "base_model": BASE_MODEL,
    "seed": SEED,
    "train_size": int(len(train_df)),
    "eval_size": int(len(eval_df)),
    "batch_size": BATCH_SIZE,
    "num_epochs": NUM_EPOCHS,
    "num_iterations": NUM_ITERATIONS,
    "eval_accuracy": float(acc),
    "train_duration_minutes": float(train_minutes),
}

# Save metrics + report
with open(FINETUNE_DIR / "metrics_eval.json", "w", encoding="utf-8") as f:
    json.dump({"metrics": metrics, "classification_report": report_dict}, f, ensure_ascii=False, indent=2)

report_df = pd.DataFrame(report_dict).transpose()
report_df.to_csv(FINETUNE_DIR / "classification_report_eval.csv", index=True)
```

### 4.1.1 - Strategic Data Selection for Active Learning Round 1

```python
# 0) CONFIG
PROJECT_ROOT = Path(".").resolve()

# Unlabeled pool (baseline analyzed 4623)
POOL_PATH = PROJECT_ROOT / "Output Analysis" / "Sentiment Analysis Dataset" / "Sentiment_Analyzed.csv"

# Gold dataset (untuk exclude dari pool)
GOLD_PATH = PROJECT_ROOT / "Output Analysis" / "Baseline Dataset" / "baseline_labeling_sample_done.csv"

# Folder fine-tune run (mengambil latest run otomatis)
FINETUNE_ROOT = PROJECT_ROOT / "artifacts" / "setfit_finetune"
latest_run = sorted(FINETUNE_ROOT.glob("run_*"))[-1]
MODEL_DIR = latest_run / "final_model"

VALID_LABELS = ["negative", "neutral", "positive"]
SEED = 42
np.random.seed(SEED)

# Target jumlah data untuk label Round_1
N_LABEL_ROUND1 = 300

# Fixed quota selection
N_UNC = 170
N_POS = 110
N_DIS = 20

assert N_UNC + N_POS + N_DIS == N_LABEL_ROUND1

# Artifacts output
RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")
AL_DIR = PROJECT_ROOT / "artifacts" / "active_learning" / f"round_1_run_{RUN_ID}"
AL_DIR.mkdir(parents=True, exist_ok=True)

OUT_LABEL_PATH = AL_DIR / "to_label_round_1.csv"
META_PATH = AL_DIR / "round_1_selection_meta.json"

print("Loaded model from:", MODEL_DIR)
print("Pool path:", POOL_PATH)
print("Gold path:", GOLD_PATH)
print("AL artifacts dir:", AL_DIR)
```

Konfigurasi Active Learning Round_1 di atas adalah proses memilih subset komentar paling informatif dari data unlabeled untuk dilabeli manual, lalu digunakan untuk meningkatkan performa model (terutama pada kasus sulit dan kelas minoritas).

1. Menentukan Sumber Data yang Dipakai
    - Unlabeled Pool (4623 komentar) file `Sentiment_Analyzed.csv`, ini adalah dataset besar hasil baseline (IndoBERT + keyword boosting) yang belum punya label manual dan menjadi "kandidat” untuk dipilih.
    - Gold Dataset (label manual) file ` baseline_labeling_sample_done.csv`, digunakan untuk exclude komentar yang sudah pernah dilabeli agar tidak terpilih lagi (menghindari duplikasi dan data leakage).

2. Mengambil Model Fine-Tuned sebelumnya (Round_0), model tersebut dipakai sebagai “penilai” untuk menentukan komentar mana yang:
    - paling membingungkan,
	- berpotensi memperbaiki performa,
	- atau menunjukkan disagreement.
    
3. Menentukan Target Label Round_1 = 300, artinya kita menargetkan 300 komentar baru untuk diberi label manual pada Round_1 dan 300 label baru dipilih sebagai trade off optimal antara effort labeling dan expected improvement karena bukan random lagi.

4. Strategic Selection dengan Fixed Quota, Agar Round_1 tidak bias ke satu tipe data saja, data dipilih dengan komposisi tetap, yaitu:
    - 170 Uncertainty Sampling, fokus memperbaiki decision boundary (data yang paling “ragu-ragu” menurut model)
    - 110 Positive Mining, fokus memperkuat kelas positive yang minoritas dan sebelumnya lemah di Round_0
    - 20 Disagreement Cases, mengambil “hard cases” yang prediksinya berbeda antar pendekatan/model untuk memperbaiki konsistensi

```python
# 1) LOAD MODEL
if not MODEL_DIR.exists():
    raise FileNotFoundError(f"Model dir tidak ditemukan: {MODEL_DIR}")

model = SetFitModel.from_pretrained(str(MODEL_DIR))

# 2) LOAD POOL + GOLD, ANTI-LEAKAGE
if not POOL_PATH.exists():
    raise FileNotFoundError(f"Pool file tidak ditemukan: {POOL_PATH}")

df_pool = pd.read_csv(POOL_PATH)
print("\nPool loaded:", len(df_pool))

required_pool_cols = ["clean_text", "sentiment_label", "sentiment_score", "source_channel"]
missing = [c for c in required_pool_cols if c not in df_pool.columns]
if missing:
    raise ValueError(f"Pool missing columns: {missing}. Kolom tersedia: {list(df_pool.columns)}")

# Membersihkan teks
df_pool["clean_text"] = df_pool["clean_text"].astype(str).str.strip()
df_pool = df_pool[df_pool["clean_text"] != ""].copy()

# load gold untuk exclude (anti data leakage)
if not GOLD_PATH.exists():
    raise FileNotFoundError(f"Gold file tidak ditemukan: {GOLD_PATH}")

df_gold = pd.read_csv(GOLD_PATH)

# detect kolom text gold
possible_text_cols = ["clean_text", "text", "comment", "raw_text"]
gold_text_col = next((c for c in possible_text_cols if c in df_gold.columns), None)
if gold_text_col is None:
    raise ValueError(f"Kolom text di gold tidak ketemu. Kolom tersedia: {list(df_gold.columns)}")

df_gold["gold_text"] = df_gold[gold_text_col].astype(str).str.strip()
df_gold = df_gold[df_gold["gold_text"] != ""].copy()

# hashing text buat dedup yang stabil
def hash_text(s: str) -> str:
    return hashlib.md5(s.encode("utf-8")).hexdigest()

df_pool["text_hash"] = df_pool["clean_text"].apply(hash_text)
df_gold["text_hash"] = df_gold["gold_text"].apply(hash_text)

gold_hashes = set(df_gold["text_hash"].unique().tolist())

before = len(df_pool)
df_pool = df_pool[~df_pool["text_hash"].isin(gold_hashes)].copy()
after = len(df_pool)

print(f"Anti-leakage exclude gold: sebelum={before} sesudah={after} drop={before-after}")
```

Pada tahap ini, sistem memuat model hasil fine-tuning Round_0 serta dataset pool yang berisi 4.623 komentar hasil analisis baseline. Dataset pool ini merupakan sumber kandidat untuk Active Learning Round_1. Namun sebelum proses seleksi dilakukan, perlu dipastikan bahwa tidak ada komentar yang sudah pernah dilabeli sebelumnya ikut terpilih kembali.

Untuk menjaga integritas eksperimen, dataset gold (data berlabel manual pada tahap sebelumnya) dimuat dan dibandingkan dengan pool menggunakan teknik hashing pada teks yang sudah dibersihkan. Setiap komentar diubah menjadi representasi hash unik (MD5), sehingga proses pencocokan tidak bergantung pada urutan baris atau format teks, melainkan pada isi aktualnya. Dengan cara ini, sistem dapat mengidentifikasi komentar yang identik secara akurat dan stabil.

Hasilnya menunjukkan bahwa dari 4.623 komentar awal dalam pool, sebanyak 697 komentar terdeteksi sudah pernah digunakan dalam dataset gold dan kemudian dikeluarkan. Setelah proses anti-leakage ini, tersisa 3.926 komentar yang benar-benar baru dan layak dijadikan kandidat untuk seleksi Active Learning Round_1.

Langkah ini sangat penting karena mencegah data leakage, yaitu kondisi ketika model kembali belajar dari data yang pernah digunakan sebelumnya. Tanpa proses ini, peningkatan performa model pada round berikutnya bisa tampak lebih baik secara angka, tetapi sebenarnya tidak mencerminkan kemampuan generalisasi yang sesungguhnya. Dengan menghapus overlap tersebut, setiap peningkatan performa pada Round_1 benar-benar berasal dari pembelajaran terhadap data baru, sehingga hasil eksperimen tetap valid dan dapat dipertanggungjawabkan secara ilmiah.

```python
# 3) SCORE POOL WITH PROBA + UNCERTAINTY
texts = df_pool["clean_text"].tolist()

# predict_proba: shape (n_samples, n_labels)
proba = model.predict_proba(texts)
proba = np.array(proba)

# pastikan urutan label konsisten dengan VALID_LABELS (SetFit pakai labels dari model)
# kita asumsikan model.labels == VALID_LABELS (seperti waktu training diatas)
# kalau ragu, bisa print(model.labels)
if hasattr(model, "labels"):
    model_labels = list(model.labels)
else:
    model_labels = VALID_LABELS

# mapping index untuk ambil kolom proba sesuai label target
idx_map = {lab: model_labels.index(lab) for lab in VALID_LABELS}

p_neg = proba[:, idx_map["negative"]]
p_neu = proba[:, idx_map["neutral"]]
p_pos = proba[:, idx_map["positive"]]

pred_idx = proba.argmax(axis=1)
pred_label = [model_labels[i] for i in pred_idx]

# entropy (uncertainty): makin tinggi makin bingung
eps = 1e-12
entropy = -(proba * np.log(proba + eps)).sum(axis=1)

# margin (selisih top1 vs top2): makin kecil makin bingung
sorted_proba = np.sort(proba, axis=1)
margin = sorted_proba[:, -1] - sorted_proba[:, -2]

df_pool["pred_label_setfit"] = pred_label
df_pool["p_negative"] = p_neg
df_pool["p_neutral"]  = p_neu
df_pool["p_positive"] = p_pos
df_pool["uncertainty_entropy"] = entropy
df_pool["uncertainty_margin"] = margin
```

```python
# 4) STRATEGIC SELECTION (FIXED QUOTA)

N_UNC = 170
N_POS = 110
N_DIS = 20

# Uncertainty
cand_unc = (
    df_pool.sort_values(["uncertainty_entropy", "uncertainty_margin"], ascending=[False, True])
    .head(N_UNC)
    .copy()
)
cand_unc["selection_reason"] = "uncertainty"

# Positive mining
cand_pos = df_pool.copy()
cand_pos["pos_rank_score"] = (
    0.75 * cand_pos["p_positive"] +
    0.25 * cand_pos["uncertainty_entropy"]
)

cand_pos = (
    cand_pos.sort_values("pos_rank_score", ascending=False)
    .loc[~cand_pos["text_hash"].isin(cand_unc["text_hash"])]
    .head(N_POS)
    .copy()
)
cand_pos["selection_reason"] = "positive_mining"

# Disagreement
df_pool["sentiment_label"] = df_pool["sentiment_label"].astype(str).str.lower().str.strip()

disagree = df_pool[
    (df_pool["sentiment_label"].isin(VALID_LABELS)) &
    (df_pool["sentiment_label"] != df_pool["pred_label_setfit"])
].copy()

disagree["setfit_conf"] = disagree[["p_negative", "p_neutral", "p_positive"]].max(axis=1)

disagree = (
    disagree.sort_values(["setfit_conf", "uncertainty_entropy"], ascending=[False, False])
    .loc[~disagree["text_hash"].isin(pd.concat([cand_unc, cand_pos])["text_hash"])]
    .head(N_DIS)
    .copy()
)
disagree["selection_reason"] = "disagreement"

# Digabung
df_sel = pd.concat([cand_unc, cand_pos, disagree], ignore_index=True)
df_sel = df_sel.sample(frac=1.0, random_state=SEED).reset_index(drop=True)

print("Final selected:", len(df_sel))
print(df_sel["selection_reason"].value_counts())
```

Bagian kode ini adalah inti dari Active Learning Round_1, yaitu memilih 300 komentar paling “bernilai” untuk dilabeli manual dengan strategi yang terarah, bukan sampling acak. Tujuannya adalah memaksimalkan peningkatan performa model (khususnya pada kasus yang sulit dan kelas minoritas) dengan effort labeling yang tetap realistis.

Seleksi dilakukan menggunakan fixed quota agar komposisi data yang diambil seimbang secara strategi: 170 data dipilih dari komentar yang paling tidak pasti menurut model (uncertainty sampling), 110 data dipilih dari komentar yang memiliki probabilitas tinggi sebagai positif namun tetap mengandung unsur ketidakpastian (positive mining), dan 20 data dipilih dari kasus “sulit” di mana prediksi baseline berbeda dengan prediksi SetFit (disagreement cases). Dengan pembagian ini, round_1 tidak hanya memperbaiki decision boundary secara umum, tetapi juga secara spesifik menambal kelemahan model pada kelas positif yang sebelumnya minoritas dan sering jatuh ke netral.

Secara mekanisme, kandidat uncertainty diambil dari komentar dengan tingkat ketidakpastian tertinggi (misalnya entropy tinggi), karena komentar seperti ini paling informatif untuk memperjelas batas antar kelas. Setelah itu, positive mining memberi prioritas pada komentar yang kemungkinan besar positif (berdasarkan p_positive), namun masih cukup “menantang” (ditambah bobot uncertainty) sehingga bagus untuk memperkaya variasi positive dan mengurangi bias model ke kelas netral. Terakhir, disagreement memilih komentar yang label baseline-nya berbeda dengan prediksi SetFit—ini biasanya merupakan hard cases yang rawan salah klasifikasi, sehingga labeling manual pada bagian ini membantu meningkatkan konsistensi model.

Hasil akhirnya menunjukkan seleksi berjalan sesuai desain: tepat 300 komentar terpilih dengan komposisi 170/110/20 sesuai rencana. Setelah digabung, data diacak agar urutan saat labeling tidak bias terhadap satu jenis kasus tertentu. Dengan pendekatan ini, Round_1 menjadi lebih efisien karena setiap komentar yang dilabeli dipilih untuk memberi “nilai belajar” maksimal bagi model.

```python
# 3) NORMALIZE COLUMNS -> keep only text + label (SetFit needs text,label)
possible_text_cols  = ["clean_text", "text", "comment", "raw_text"]
possible_label_cols = ["gold_label", "label", "sentiment_label", "sentiment"]

def normalize_gold(df, df_name="df"):
    text_col  = next((c for c in possible_text_cols  if c in df.columns), None)
    label_col = next((c for c in possible_label_cols if c in df.columns), None)

    if text_col is None or label_col is None:
        raise ValueError(
            f"[{df_name}] Kolom text/label tidak ketemu.\n"
            f"Kolom tersedia: {list(df.columns)}\n"
            f"Text candidates: {possible_text_cols}\n"
            f"Label candidates: {possible_label_cols}"
        )

    out = df[[text_col, label_col]].copy()
    out = out.rename(columns={text_col: "text", label_col: "label"})
    out["text"] = out["text"].astype(str).str.strip()
    out["label"] = out["label"].astype(str).str.lower().str.strip()
    out = out[(out["text"] != "") & (out["label"].isin(VALID_LABELS))].copy()
    return out

df_gold0_n = normalize_gold(df_gold0, "gold_v0")
df_r1_n = normalize_gold(df_r1, "round_1_done")

print("\nAfter normalize/filter:")
print("gold_v0 usable:", len(df_gold0_n))
print("round_1 usable:", len(df_r1_n))
```

```python
# 4) MERGE -> GOLD V1 (DEDUP BY TEXT HASH)
def hash_text(s: str) -> str:
    return hashlib.md5(s.encode("utf-8")).hexdigest()

df_gold0_n["text_hash"] = df_gold0_n["text"].apply(hash_text)
df_r1_n["text_hash"] = df_r1_n["text"].apply(hash_text)

# Gabung: gold_v0 dulu, lalu round_1; keep last = round_1 menang kalau ada duplikat
df_merged = pd.concat([df_gold0_n, df_r1_n], ignore_index=True)
df_merged = df_merged.drop_duplicates(subset=["text_hash"], keep="last").copy()

# drop helper
df_merged = df_merged.drop(columns=["text_hash"])

# Save gold_v1
df_merged.to_csv(GOLD_V1_PATH, index=False)

print("\nSaved gold_v1:", GOLD_V1_PATH)
print("gold_v1 size:", len(df_merged))
print("\nDistribusi label (gold_v1):")
print(df_merged["label"].value_counts())

# Save merge meta (nyambung artifacts)
merge_meta = {
    "run_id": RUN_ID,
    "timestamp": datetime.now().isoformat(),
    "gold_v0_path": str(GOLD_V0_PATH),
    "round1_done_path": str(ROUND1_DONE_PATH),
    "gold_v1_output_path": str(GOLD_V1_PATH),
    "n_gold_v0_usable": int(len(df_gold0_n)),
    "n_round1_usable": int(len(df_r1_n)),
    "n_gold_v1_final": int(len(df_merged)),
    "notes": "gold_v1 = gold_v0 + round_1 labeled (dedup by text hash, keep last=round_1)"
}
with open(FINETUNE_DIR / "merge_meta.json", "w", encoding="utf-8") as f:
    json.dump(merge_meta, f, ensure_ascii=False, indent=2)
```

```python
# 5) SPLIT TRAIN/EVAL (STRATIFIED)
do_stratify = df_merged["label"].nunique() > 1 and df_merged["label"].value_counts().min() >= 2

train_df, eval_df = train_test_split(
    df_merged,
    test_size=0.2,
    random_state=SEED,
    stratify=df_merged["label"] if do_stratify else None
)

# Save label distribution snapshot
dist_train = train_df["label"].value_counts()
dist_eval  = eval_df["label"].value_counts()
dist_df = pd.DataFrame({"train": dist_train, "eval": dist_eval}).fillna(0).astype(int)
dist_df.to_csv(FINETUNE_DIR / "label_distribution_train_eval.csv")

print(f"\nTrain: {len(train_df)} | Eval: {len(eval_df)} | Stratify: {do_stratify}")

train_dataset = Dataset.from_pandas(train_df.reset_index(drop=True), preserve_index=False)
eval_dataset  = Dataset.from_pandas(eval_df.reset_index(drop=True), preserve_index=False)
```

```python
# 6) LOAD MODEL
model = SetFitModel.from_pretrained(
    BASE_MODEL,
    labels=VALID_LABELS
)

# Optional move to device (safe)
try:
    model.model_body.to(device)
except Exception:
    pass
if getattr(model, "model_head", None) is not None:
    try:
        model.model_head.to(device)
    except Exception:
        pass
```

```python
# 7) TRAINING CONFIG
if device == "cuda":
    NUM_ITERATIONS = 12
    BATCH_SIZE     = 32
    NUM_EPOCHS     = 3
else:
    NUM_ITERATIONS = 6
    BATCH_SIZE     = 16
    NUM_EPOCHS     = 2

args = SetFitTrainingArguments(
    batch_size=BATCH_SIZE,
    num_epochs=NUM_EPOCHS,
    num_iterations=NUM_ITERATIONS,
)

args.eval_strategy = "epoch"
args.save_strategy = "epoch"
args.load_best_model_at_end = True

# If supported by this version
if hasattr(args, "output_dir"):
    args.output_dir = str(CHECKPOINT_DIR)

trainer = SetFitTrainer(
    model=model,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

print("\nTraining SetFit (Round_1)...")
print(f"Config -> device={device}, batch_size={BATCH_SIZE}, num_epochs={NUM_EPOCHS}, num_iterations={NUM_ITERATIONS}")
print("Checkpoint dir:", CHECKPOINT_DIR)
print("Final model dir:", FINAL_DIR)

t0 = time.time()
trainer.train()
t1 = time.time()
train_minutes = (t1 - t0) / 60

print(f"\nTraining selesai. Durasi: {train_minutes:.2f} menit")
```

### 4.1.3 - Fixed Test Set Evaluation (Round 1 & Round 2)

`Fixed Test Set` adalah sekumpulan data yang dipisahkan sejak awal dan tidak pernah digunakan untuk training atau active learning. Dataset ini berfungsi sebagai “soal ujian tetap” untuk mengukur performa model secara konsisten di setiap iterasi (Round_0 dan Round_1).

Sebelumnya, metrik Round_1 terlihat turun karena dataset evaluasi berubah (split ulang setelah data bertambah). Artinya, model diuji pada data yang berbeda dari Round_0, sehingga perbandingan menjadi tidak adil. Inilah mengapa fixed test diperlukan.

Dengan pendekatan ini:
- Perbandingan antar round menjadi adil (apples to apples).
- Jika skor naik, berarti model benar-benar membaik.
- Jika skor turun, berarti ada masalah generalisasi.
- Tidak ada bias akibat perubahan distribusi data.

Singkatnya, fixed test set memastikan bahwa peningkatan performa bukan karena “soal ujiannya berubah”, tetapi karena model memang lebih baik.

```python
# 1) Baseline predictions on fixed test (join from Sentiment_Analyzed.csv)
df_base = pd.read_csv(BASELINE_ANALYZED_PATH).copy()
df_base["clean_text"] = df_base["clean_text"].astype(str).str.strip()
df_base["sentiment_label"] = df_base["sentiment_label"].astype(str).str.lower().str.strip()
df_base = df_base[df_base["sentiment_label"].isin(VALID_LABELS)].copy()

df_base["text_hash"] = df_base["clean_text"].apply(hash_text)

# mapping hash -> baseline pred
base_map = df_base.drop_duplicates("text_hash")[["text_hash", "sentiment_label"]].set_index("text_hash")["sentiment_label"]

df_test["pred_baseline"] = df_test["text_hash"].map(base_map)

# cek coverage baseline
coverage = df_test["pred_baseline"].notna().mean()
missing = df_test["pred_baseline"].isna().sum()
print(f"\nBaseline join coverage: {coverage*100:.1f}% | missing: {missing}")

# jika ada missing, kita drop baris itu untuk fairness baseline
df_eval_base = df_test[df_test["pred_baseline"].notna()].copy()
y_true_base = df_eval_base["label"].tolist()
y_pred_base = df_eval_base["pred_baseline"].tolist()
```

```python
# 2) Round_0 & Round_1 predictions (SetFit)
model_r0 = SetFitModel.from_pretrained(str(MODEL_R0_DIR))
model_r1 = SetFitModel.from_pretrained(str(MODEL_R1_DIR))

y_pred_r0 = model_r0.predict(df_test["text"].tolist())
y_pred_r1 = model_r1.predict(df_test["text"].tolist())

# Helper to compute metrics
def summarize_metrics(name, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    rep = classification_report(y_true, y_pred, output_dict=True, digits=4, zero_division=0)

    out = {
        "model": name,
        "accuracy": acc,
        "macro_f1": rep["macro avg"]["f1-score"],
        "weighted_f1": rep["weighted avg"]["f1-score"],
        "f1_negative": rep.get("negative", {}).get("f1-score", np.nan),
        "f1_neutral": rep.get("neutral", {}).get("f1-score", np.nan),
        "f1_positive": rep.get("positive", {}).get("f1-score", np.nan),
        "support_total": sum(rep[c]["support"] for c in ["negative","neutral","positive"] if c in rep),
    }
    return out, rep

rows = []
reports = {}

# Baseline (pakai subset yang match)
row_base, rep_base = summarize_metrics("Baseline Hybrid", y_true_base, y_pred_base)
rows.append(row_base)
reports["Baseline Hybrid"] = rep_base

# Round_0 & Round_1 (full fixed test)
row_r0, rep_r0 = summarize_metrics("SetFit Round_0", y_true, y_pred_r0)
row_r1, rep_r1 = summarize_metrics("SetFit Round_1", y_true, y_pred_r1)

rows.append(row_r0)
rows.append(row_r1)

reports["SetFit Round_0"] = rep_r0
reports["SetFit Round_1"] = rep_r1

df_cmp = pd.DataFrame(rows)

# Save comparison table
SUMMARY_DIR = PROJECT_ROOT / "artifacts" / "summary"
SUMMARY_DIR.mkdir(parents=True, exist_ok=True)

cmp_path = SUMMARY_DIR / "comparison_baseline_vs_round0_vs_round1_fixedtest.csv"
df_cmp.to_csv(cmp_path, index=False)

print("\n✅ Comparison table saved:", cmp_path)
display(df_cmp)

# Save classification reports per model
for k, rep in reports.items():
    out_path = SUMMARY_DIR / f"classification_report_{k.lower().replace(' ','_')}.json"
    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(rep, f, ensure_ascii=False, indent=2)
```

## **4.  Aspect-Based Categorization (Otomatisasi Topik)**

Setelah model final SetFit selesai dan performanya tervalidasi menggunakan fixed test set, langkah berikutnya  adalah mengidentifikasi topik utama yang mendorong sentimen tersebut. 

Melalui `Aspect-Based Categorization`, komentar pengguna secara otomatis dikelompokkan ke dalam aspek produk seperti `kamera`, `baterai`, `performa` dll, sehingga analisis tidak hanya menjawab “bagaimana persepsi pasar”, tetapi juga “terhadap fitur apa persepsi tersebut terbentuk”.

Jika sentiment analysis menjawab emosi pasar, maka aspect-based categorization menjawab topik pembicaraan pasar. Tanpa aspek, kita hanya tahu 30% negatif. Dengan aspek, kita tahu 30% negatif itu karena kamera, baterai, atau software.

Tahap ini bertujuan untuk:
- Mengotomatisasi pengelompokan topik komentar
- Menghubungkan sentimen dengan aspek produk
- Mengubah data opini menjadi insight yang bisa ditindaklanjuti (actionable)

```python
# 0) LOAD DATA FINAL SETFIT
df = pd.read_csv(
    "Output Analysis/Sentiment Analysis Dataset/Sentiment_Analyzed_setfit_final.csv"
)

# 1) VALIDASI KOLOM
if "clean_text" not in df.columns:
    raise ValueError("Kolom `clean_text` tidak ada.")

LABEL_COL_FINAL = "sentiment_label_final"

if LABEL_COL_FINAL not in df.columns:
    raise ValueError(
        f"Kolom `{LABEL_COL_FINAL}` tidak ditemukan. Pastikan ini file output final SetFit."
    )

# 2) BUAT POLARITY DARI LABEL FINAL
label_to_polarity = {"negative": -1, "neutral": 0, "positive": 1}

df[LABEL_COL_FINAL] = df[LABEL_COL_FINAL].astype(str).str.lower().str.strip()
df["polarity"] = df[LABEL_COL_FINAL].map(label_to_polarity)

# Filter hanya label valid
df = df[df["polarity"].isin([-1, 0, 1])].copy()
```

```python
ASPECTS_DICT = {

    # KAMERA
    "kamera": [
        "kamera", "cam", "leica", "kamera leica", "boba",
        "foto", "video", "rekam",
        "selfie", "kamera depan", "kamera belakang",
        "night", "night mode", "low light", "malam",
        "zoom", "tele", "telephoto", "ultrawide",
        "potret", "portrait", "bokeh",
        "ois", "eis",
        "4k", "8k",
        "200mp", "108mp", "50mp"
    ],

    # BATERAI
    "baterai": [
        "baterai", "battery", "batre", "batrei",
        "daya tahan", "kapasitas",
        "5000mah", "5500mah", "6000mah",
        "screen on time", "sot",
        "tahan seharian", "dipakai seharian"
    ],

    # CHARGING
    "charging": [
        "charge", "charging", "cas", "ngecas",
        "fast charging", "pengisian cepat",
        "hypercharge", "turbo charge",
        "67w", "90w", "120w",
        "wireless charging", "reverse charging"
    ],

    # PANAS / THERMAL
    "thermal": [
        "panas", "overheat", "hangat", "gerah",
        "thermal", "throttling",
        "panas gaming", "panas main game",
        "panas saat cas",
        "cooling", "vapor chamber", "liquid cool"
    ],

    # PERFORMA
    "performa": [
        "performa", "kinerja",
        "chipset", "soc", "processor",
        "snapdragon", "dimensity",
        "antutu", "geekbench",
        "gaming", "fps", "frame drop",
        "multitasking",
        "pubg", "genshin", "mobile legends", "ml", "cod"
    ],

    # SOFTWARE
    "software": [
        "hyperos", "miui",
        "software", "sistem",
        "ui", "antarmuka",
        "update", "pembaruan",
        "bug", "error",
        "bloatware", "iklan",
        "fitur", "gesture", "animasi"
    ],

    # LAYAR
    "layar": [
        "layar", "screen",
        "amoled", "oled", "ltpo",
        "refresh rate", "hz", "120hz", "144hz",
        "resolusi", "1.5k", "2k",
        "brightness", "nits",
        "hdr", "dolby vision",
        "fingerprint", "sidik jari",
        "pwm", "eye care"
    ],

    # DESAIN & BUILD
    "desain": [
        "desain", "design",
        "build", "build quality",
        "bodi", "body", "frame",
        "material",
        "kaca", "glass",
        "flat", "curve", "curved",
        "bezel", "tebal", "tipis",
        "warna", "color", "finishing"
    ],

    # HARGA & VALUE
    "harga": [
        "harga", "price",
        "jutaan", "rp",
        "murah", "mahal",
        "worth", "worth it",
        "value", "value for money",
        "promo", "diskon", "flash sale",
        "shopee", "tokopedia", "lazada"
    ],
}
```

```python
# 4) COMPILE REGEX (MINIMAL IMPROVE)
ASPECT_PATTERNS = {}

for aspect, keywords in ASPECTS_DICT.items():
    patterns = []
    seen = set()

    for kw in keywords:
        kw = kw.strip().lower()
        if not kw or kw in seen:
            continue

        seen.add(kw)

        if len(kw) <= 2:
            patterns.append(re.compile(rf"\b{re.escape(kw)}\b"))
        else:
            patterns.append(re.compile(rf"(?<!\w){re.escape(kw)}(?!\w)"))

    ASPECT_PATTERNS[aspect] = patterns

# 5) TAGGING FUNCTION
def tag_aspects(text: str):
    if not isinstance(text, str) or not text.strip():
        return ["umum"]

    t = text.lower()
    tags = []

    for aspect, patterns in ASPECT_PATTERNS.items():
        if any(p.search(t) for p in patterns):
            tags.append(aspect)

    return tags if tags else ["umum"]

# 6) TAGGING + EXPLODE
df["aspects"] = df["clean_text"].astype(str).apply(tag_aspects)

df_exploded = df.explode("aspects")
```

```python
# 7) SUMMARY PER ASPEK
aspect_summary = (
    df_exploded
    .groupby("aspects", dropna=False)
    .agg(
        total_mentions=("clean_text", "count"),
        avg_sentiment=("polarity", "mean"),
    )
    .sort_values("total_mentions", ascending=False)
)
```

Setelah kita menghitung rata-rata sentimen (avg_sentiment) untuk setiap aspek, muncul satu pertanyaan penting: bagaimana cara menentukan apakah suatu aspek benar-benar kuat atau lemah dibanding aspek lainnya? Rata-rata sentimen saja belum cukup menjawab itu. Sebuah aspek bisa saja memiliki nilai 0.15 dan terlihat “cukup positif”, tetapi jika rata-rata keseluruhan aspek berada di 0.40, maka 0.15 sebenarnya relatif rendah.

Di sinilah Z-score digunakan.

`Z-score` adalah teknik normalisasi statistik yang mengukur seberapa jauh suatu nilai menyimpang dari rata-rata keseluruhan dalam satuan standar deviasi. Dengan kata lain, Z-score tidak melihat nilai absolut, tetapi posisi relatif suatu aspek dibandingkan aspek lain dalam sistem yang sama.

Untuk langkah pertama yaitu menghitung rata-rata (mean) dan standar deviasi (std) dari seluruh nilai avg_sentiment antar aspek. Rata-rata ini menjadi titik referensi sentimen umum produk, sedangkan standar deviasi menggambarkan seberapa besar variasi sentimen antar aspek.

Kemudian setiap aspek dihitung Z-score-nya menggunakan rumus:

*Z = (avg_sentiment - mean)/std*

Jika suatu aspek memiliki Z-score positif, berarti sentimennya lebih tinggi dari rata-rata keseluruhan aspek. Jika Z-score negatif, berarti sentimennya lebih rendah dari rata-rata. Semakin besar nilai absolutnya, semakin signifikan perbedaannya.

Setelah Z-score dihitung, aspek dikategorikan secara interpretatif menjadi tiga status:
- 🟢 Positif jika Z-score lebih dari 0.5
- 🔴 Negatif jika Z-score kurang dari -0.5
- 🟡 Netral jika berada di antara keduanya

Ambang ±0.5 digunakan agar klasifikasi tidak terlalu sensitif terhadap perbedaan kecil yang tidak signifikan secara strategis.

Pendekatan ini penting karena memungkinkan kita melihat aspek secara relatif, bukan absolut. Misalnya, semua aspek mungkin memiliki nilai sentimen positif, tetapi tetap ada aspek yang “kurang positif” dibanding yang lain. Dalam konteks bisnis, aspek tersebut bisa menjadi area yang perlu ditingkatkan meskipun tidak terlihat negatif secara langsung.

Dengan Z-score, analisis menjadi lebih tajam. Kita tidak hanya mengetahui bagaimana sentimen suatu aspek, tetapi juga seberapa kuat atau lemahnya aspek tersebut dibandingkan fitur lainnya dalam persepsi pasar. Ini memberikan dasar yang lebih objektif untuk menentukan prioritas perbaikan maupun penguatan dalam strategi produk.

```python
# 8) Z-SCORE (BASED ON AVG_SENTIMENT)
mean = aspect_summary["avg_sentiment"].mean()
std = aspect_summary["avg_sentiment"].std()

if std == 0 or np.isnan(std):
    aspect_summary["z_score"] = 0.0
else:
    aspect_summary["z_score"] = (
        (aspect_summary["avg_sentiment"] - mean) / std
    )

aspect_summary["sentiment_status"] = aspect_summary["z_score"].apply(
    lambda z: "🟢 Positif" if z > 0.5
    else ("🔴 Negatif" if z < -0.5 else "🟡 Netral")
)

print("SENTIMEN BERDASARKAN ASPEK (Z-SCORE) — FINAL SETFIT MODEL")
display(aspect_summary)
```

Hasil ini menunjukkan bagaimana persepsi pasar terdistribusi pada level fitur setelah dinormalisasi menggunakan Z-score. Artinya, kita tidak lagi melihat sentimen secara absolut, tetapi secara relatif dibandingkan antar aspek.

Secara umum, aspek “umum” memiliki jumlah mention paling besar (3455) dan dikategorikan sebagai 🟢 Positif berdasarkan Z-score. Ini menunjukkan bahwa mayoritas komentar yang tidak secara spesifik menyebut fitur tertentu cenderung bernada cukup baik. Namun karena ini kategori general, insight strategisnya lebih terbatas dibanding aspek teknis.

Dua aspek yang paling menonjol secara negatif adalah desain dan kamera.
Desain memiliki Z-score paling rendah (-2.17) dengan rata-rata sentimen -0.76, menjadikannya aspek dengan deviasi negatif paling kuat dibanding aspek lain. Ini mengindikasikan bahwa persepsi terhadap desain bukan sekadar sedikit negatif, tetapi secara statistik jauh lebih buruk dibanding rata-rata keseluruhan aspek.

Kamera juga berada di zona 🔴 Negatif dengan Z-score -1.30 dan volume pembicaraan cukup tinggi (468 mention). Ini signifikan secara bisnis karena kamera adalah fitur utama di segmen mid-range. Artinya, ini bukan hanya isu minor, tetapi berpotensi memengaruhi positioning produk.

Beberapa aspek berada di zona 🟡 Netral seperti:
- Harga
- Layar
- Software
- Performa

Artinya, persepsi pasar terhadap fitur-fitur ini relatif seimbang dibanding aspek lain. Tidak menjadi pain point besar, tetapi juga belum menjadi selling point dominan.

Menariknya, beberapa aspek justru relatif positif:
- Baterai (Z-score 0.59)
- Thermal (0.54)
- Charging (1.10)

Charging memiliki Z-score paling tinggi di antara aspek spesifik, meskipun volumenya kecil. Ini menunjukkan bahwa bagi pengguna yang membahas charging, persepsinya cenderung jauh lebih baik dibanding rata-rata aspek lain.

Namun perlu dicatat bahwa aspek dengan total mention rendah (misalnya charging dan thermal) perlu dibaca hati-hati. Walaupun secara relatif positif, volumenya kecil sehingga dampak strategisnya tidak sebesar aspek dengan volume tinggi seperti kamera atau desain.

⸻

Insight Strategis Utama
1.	Critical Weakness
	- Desain dan kamera adalah area paling membutuhkan perhatian.
	- Kamera terutama penting karena volumenya besar dan sentimennya relatif rendah.
2.	Stable / Competitive Areas
	- Layar, software, performa, harga berada di zona netral.
	- Tidak menjadi sumber keluhan ekstrem.
3.	Emerging Strengths
	- Charging dan baterai menunjukkan persepsi relatif baik.
	- Potensi untuk di-highlight dalam komunikasi marketing.

```python
LABEL_ORDER = ["negative", "neutral", "positive"]
LABEL_COL_FINAL = "sentiment_label_final"

# VALIDATION GUARD
if "aspects" not in df_exploded.columns:
    raise ValueError("df_exploded belum punya kolom `aspects`. Jalankan tagging dulu.")

if LABEL_COL_FINAL not in df_exploded.columns:
    raise ValueError(
        f"Kolom `{LABEL_COL_FINAL}` tidak ada. Pastikan ini output final SetFit."
    )

# normalisasi label final (jaga-jaga)
df_exploded[LABEL_COL_FINAL] = (
    df_exploded[LABEL_COL_FINAL]
    .astype(str)
    .str.lower()
    .str.strip()
)

df_exploded = df_exploded[
    df_exploded[LABEL_COL_FINAL].isin(LABEL_ORDER)
].copy()

# COUNT PER ASPEK x LABEL
aspect_label_counts = (
    df_exploded
    .pivot_table(
        index="aspects",
        columns=LABEL_COL_FINAL,
        values="clean_text",
        aggfunc="count",
        fill_value=0
    )
    .reindex(columns=LABEL_ORDER, fill_value=0)
)

# TOTAL + RATE
aspect_label_counts["total_mentions"] = aspect_label_counts.sum(axis=1)

aspect_label_rates = aspect_label_counts.copy()

for lab in LABEL_ORDER:
    aspect_label_rates[f"{lab}_rate"] = np.where(
        aspect_label_rates["total_mentions"] > 0,
        aspect_label_rates[lab] / aspect_label_rates["total_mentions"],
        0.0
    )

# MERGE COUNT + RATE
aspect_distribution = (
    pd.concat(
        [
            aspect_label_counts,
            aspect_label_rates[[f"{l}_rate" for l in LABEL_ORDER]]
        ],
        axis=1
    )
    .sort_values("total_mentions", ascending=False)
)

print("DISTRIBUSI SENTIMEN PER ASPEK (FINAL SETFIT MODEL)")
display(aspect_distribution)
```

Dari tabel distribusi sentimen per aspek, menggambarkan kondisi mentah persepsi pengguna. Dari sini terlihat bahwa tidak semua fitur diperlakukan sama oleh pasar. Ada fitur yang relatif stabil dan cenderung netral, tetapi ada juga fitur yang secara konsisten memunculkan keluhan. Dua aspek paling mencolok adalah `desain` dan `kamera`. `Desain` memiliki proporsi negatif yang sangat dominan, mendekati delapan dari sepuluh komentar bernada negatif. `Kamera` juga menunjukkan pola serupa dengan lebih dari separuh pembahasannya bernada negatif, dan yang membuatnya lebih krusial adalah volumenya yang sangat tinggi. Artinya, bukan hanya banyak yang tidak puas, tetapi juga banyak yang membicarakannya. Ini adalah kombinasi yang berisiko dalam konteks positioning produk.

Aspek seperti `harga`, `layar`, dan `software` berada pada zona yang lebih sedang. Mayoritas percakapannya netral, dengan sebagian keluhan tetapi tidak dominan. Ini menunjukkan bahwa fitur-fitur tersebut belum menjadi kekuatan utama, namun juga belum menjadi sumber krisis. Sementara itu, `baterai`, `performa`, `thermal`, dan `charging` relatif stabil dengan proporsi negatif yang lebih rendah. Namun volume pembicaraannya tidak sebesar `kamera` atau `desain`, sehingga dampaknya terhadap persepsi keseluruhan tidak sebesar dua aspek utama tadi.

```python
# DISTRIBUSI SENTIMENT PER ASPEK
aspect_distribution = (
    df_exploded
    .groupby(["aspects", LABEL_COL_FINAL])
    .size()
    .unstack(fill_value=0)
)

# Memastikan semua kolom ada
for col in ["negative", "neutral", "positive"]:
    if col not in aspect_distribution.columns:
        aspect_distribution[col] = 0

# total mention per aspek
aspect_distribution["total_mentions"] = aspect_distribution.sum(axis=1)

# Menghitung rate
aspect_distribution["negative_rate"] = (
    aspect_distribution["negative"] / aspect_distribution["total_mentions"]
)

aspect_distribution["neutral_rate"] = (
    aspect_distribution["neutral"] / aspect_distribution["total_mentions"]
)

aspect_distribution["positive_rate"] = (
    aspect_distribution["positive"] / aspect_distribution["total_mentions"]
)

# PAIN POINT ANALYSIS
MIN_MENTIONS = 30  # threshold untuk fokus ke aspek yang cukup banyak dibahas

pain_points = (
    aspect_distribution
    .query("total_mentions >= @MIN_MENTIONS")
    .copy()
)

# severity x volume balance
pain_points["pain_score"] = (
    pain_points["negative_rate"] * np.log1p(pain_points["total_mentions"])
)

pain_points = (
    pain_points
    .sort_values(
        ["pain_score", "negative_rate", "total_mentions"],
        ascending=False
    )
    [[
        "total_mentions",
        "negative", "neutral", "positive",
        "negative_rate", "neutral_rate", "positive_rate",
        "pain_score"
    ]]
)

print(f"TOP PAIN POINTS (min mentions = {MIN_MENTIONS}) — FINAL SETFIT")
display(pain_points.head(5))
```

Pada tabel Top Pain Points, membawa analisis ke level prioritas strategis. Di sini, kita tidak lagi hanya melihat persentase negatif, tetapi menggabungkannya dengan volume pembicaraan untuk menghasilkan skor dampak. Hasilnya mempertegas bahwa `desain` dan `kamera` bukan hanya negatif secara proporsi, tetapi juga signifikan secara skala diskusi. `Desain` muncul sebagai pain point paling dominan secara skor, diikuti `kamera`. Ini menunjukkan bahwa jika perusahaan harus memilih area perbaikan dengan urgensi tertinggi, dua fitur inilah yang paling rasional untuk diprioritaskan.

```python
# POSITIVE VALUE PROPOSITION
MIN_MENTIONS = 30  # tetap bisa adjust

positive_points = (
    aspect_distribution
    .query("total_mentions >= @MIN_MENTIONS")
    .copy()
)

# severity x volume balance (versi positif)
positive_points["positive_score"] = (
    positive_points["positive_rate"] * np.log1p(positive_points["total_mentions"])
)

positive_points = (
    positive_points
    .sort_values(
        ["positive_score", "positive_rate", "total_mentions"],
        ascending=False
    )
    [[
        "total_mentions",
        "negative", "neutral", "positive",
        "negative_rate", "neutral_rate", "positive_rate",
        "positive_score"
    ]]
)

print(f"TOP POSITIVE VALUE PROPOSITION (min mentions = {MIN_MENTIONS}) — FINAL SETFIT")
display(positive_points.head(5))
```

Pada tabel Top Positive Value Proposition, memberi perspektif penyeimbang. Di sini terlihat bahwa `layar` dan `baterai` relatif memiliki kontribusi positif dibanding aspek lain. Artinya, ada elemen produk yang dihargai pengguna dan berpotensi menjadi fondasi komunikasi pemasaran. Namun kekuatan positif ini belum cukup besar untuk sepenuhnya mengimbangi tekanan negatif pada `desain` dan `kamera`.

>Secara keseluruhan, analisis ini tidak hanya menunjukkan mana yang positif atau negatif, tetapi juga mana yang paling berdampak secara strategis. Inilah yang membuat hasil ini relevan untuk diterjemahkan ke dalam keputusan produk, pembaruan software, dan penyesuaian positioning pasar.

### 5.3. Keunggulan Utama (Positive Aspects)

```python
# 1. ASPECT STATISTICS (POSITIVE — FINAL SETFIT)
aspect_stats_pos = (
    df_exploded
    .groupby("aspects")
    .agg(
        total_mentions=(LABEL_COL_FINAL, "count"),
        positive_mentions=(LABEL_COL_FINAL, lambda x: (x == "positive").sum())
    )
)

# Hindari division by zero
aspect_stats_pos["pos_ratio"] = np.where(
    aspect_stats_pos["total_mentions"] > 0,
    aspect_stats_pos["positive_mentions"] / aspect_stats_pos["total_mentions"],
    0
)

# 2. FILTER NOISE & RANKING
MIN_MENTIONS = 30  # konsisten dengan pain point

top_positive = (
    aspect_stats_pos
    .query("total_mentions >= @MIN_MENTIONS")
    .sort_values(["pos_ratio", "positive_mentions"], ascending=False)
    .head(5)
    .reset_index()
)

# 3. VISUALIZATION
plt.figure(figsize=(8, 4))

sns.barplot(
    data=top_positive,
    x="pos_ratio",
    y="aspects",
    hue="aspects",
    palette="Greens",
    legend=False
)

plt.xlabel("Proporsi Komentar Positif")
plt.ylabel("")
plt.title("Keunggulan Utama (Positive Aspects) — Final SetFit")

plt.xlim(0, 1)
plt.tight_layout()
plt.show()
```

### 5.5. Prioritization Matrix: Impact vs Sentiment

```python
# PRIORITIZATION MATRIX (FINAL SETFIT)
# 1. PERSIAPAN DATA
aspect_metrics = (
    df_exploded
    .groupby("aspects")
    .agg(
        frequency=("clean_text", "count"),
        avg_sentiment=("polarity", "mean")
    )
    .reset_index()
)

# Filter noise kecil
MIN_MENTIONS = 30
aspect_metrics = aspect_metrics[aspect_metrics["frequency"] >= MIN_MENTIONS].copy()

# 2. NORMALISASI SENTIMEN (Z-SCORE)
if aspect_metrics["avg_sentiment"].std() == 0:
    aspect_metrics["sentiment_z"] = 0
else:
    aspect_metrics["sentiment_z"] = zscore(aspect_metrics["avg_sentiment"])

# 3. SCATTER PLOT
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=aspect_metrics,
    x="sentiment_z",
    y="frequency",
    size="frequency",
    sizes=(200, 1200),
    hue="sentiment_z",
    palette="RdYlGn",
    edgecolor="black",
    alpha=0.85,
    legend=False
)

# 4. LABEL ASPEK
for _, row in aspect_metrics.iterrows():
    plt.text(
        row["sentiment_z"] + 0.05,
        row["frequency"],
        row["aspects"],
        fontsize=10,
        weight="bold"
    )

# 5. GARIS KUADRAN
plt.axvline(0, color="black", linestyle="--", alpha=0.6)

freq_median = aspect_metrics["frequency"].median()
plt.axhline(freq_median, color="black", linestyle="--", alpha=0.6)

# 6. ANOTASI KUADRAN
y_max = aspect_metrics["frequency"].max()

plt.text(
    -2, y_max,
    "🚨 CRITICAL ISSUES\nHigh Volume • Negative",
    color="red", fontsize=10,
    bbox=dict(facecolor="white", alpha=0.8)
)

plt.text(
    1.2, y_max,
    "💎 SELLING POINTS\nHigh Volume • Positive",
    color="green", fontsize=10,
    bbox=dict(facecolor="white", alpha=0.8)
)

plt.text(
    -2, freq_median / 2,
    "⚠ MINOR ISSUES\nLow Volume • Negative",
    color="darkred", fontsize=9,
    bbox=dict(facecolor="white", alpha=0.7)
)

plt.text(
    1.2, freq_median / 2,
    "🌱 NICE TO HAVE\nLow Volume • Positive",
    color="darkgreen", fontsize=9,
    bbox=dict(facecolor="white", alpha=0.7)
)

# 7. FINAL TOUCH
plt.title("Prioritization Matrix: Impact vs Sentiment (Final SetFit)", fontsize=16)
plt.xlabel("Sentiment Score (Z-score)")
plt.ylabel("Frequency (Volume Pembicaraan)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 8. SAVE CHART
OUTPUT_DIR = Path("artifacts/summary")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

plt.savefig(OUTPUT_DIR / "chart_matrix_priority.png", dpi=300)
```

Matrix ini merangkum seluruh analisis dalam satu peta yang mudah dipahami, sumbu horizontal menunjukkan sentimen (kiri = lebih negatif, kanan = lebih positif), dan sumbu vertikal menunjukkan seberapa sering aspek tersebut dibicarakan (volume).

Fokus utama ada di kiri atas — area High Volume & Negative. Di sini terdapat `kamera` dan `desain`. Artinya, dua fitur ini bukan hanya banyak dikeluhkan, tetapi juga sering dibahas. Singkatnya, ini adalah tahap “packaging final” yang mengubah insight data menjadi dokumen strategis yang siap dipakai dalam pengambilan keputusan.

```python
# CONFIG
OUTPUT_DIR = Path("artifacts/summary")
INPUT_TXT  = OUTPUT_DIR / "Final_Recommendation.txt"
OUTPUT_PDF = OUTPUT_DIR / "Xiaomi_Note_15_Post_Launch_Report.pdf"

# LOAD CONTENT
if not INPUT_TXT.exists():
    raise FileNotFoundError(f"File tidak ditemukan: {INPUT_TXT}")

raw_text = INPUT_TXT.read_text(encoding="utf-8").strip()
lines = [line.rstrip() for line in raw_text.split("\n")]

# PDF SETUP
doc = SimpleDocTemplate(
    str(OUTPUT_PDF),
    pagesize=A4,
    leftMargin=2.8*cm,
    rightMargin=2.8*cm,
    topMargin=3.5*cm,
    bottomMargin=2.5*cm,
)

styles = getSampleStyleSheet()

def add_style(name, **kwargs):
    if name not in styles:
        styles.add(ParagraphStyle(name=name, **kwargs))

# Cover page styles
add_style(
    "CoverMainTitle",
    fontSize=26, leading=30, alignment=TA_CENTER, spaceAfter=12,
    textColor=HexColor("#0D47A1"), fontName="Helvetica-Bold"
)
add_style(
    "CoverProduct",
    fontSize=20, leading=24, alignment=TA_CENTER, spaceAfter=0,
    textColor=HexColor("#1565C0"), fontName="Helvetica"
)

# Content styles
add_style(
    "Section",
    fontSize=15.5, leading=19, spaceBefore=16, spaceAfter=7,
    textColor=HexColor("#1565C0"), fontName="Helvetica-Bold"
)
add_style(
    "Subsection",
    fontSize=13, leading=16, spaceBefore=12, spaceAfter=6, leftIndent=8,
    textColor=HexColor("#424242"), fontName="Helvetica-Bold"
)
add_style(
    "Body",
    fontSize=10.7, leading=14.5, alignment=TA_JUSTIFY, spaceAfter=7, spaceBefore=3
)
add_style(
    "Bullet",
    fontSize=10.5, leading=14, leftIndent=26, bulletIndent=13, spaceAfter=5
)

story = []

# MARKDOWN → REPORTLAB CONVERTER
def md_to_rl(text: str) -> str:
    if not isinstance(text, str):
        return ""
    text = escape(text)
    text = re.sub(r"\*\*(.+?)\*\*", r"<b>\1</b>", text)
    # italic sengaja dimatikan supaya ga ada nested tag error
    text = text.replace("*", "")
    return text

# COVER PAGE
story.append(Spacer(1, 8*cm))
story.append(Paragraph("Post-Launch Market Intelligence Report", styles["CoverMainTitle"]))
story.append(Spacer(1, 1.2*cm))
story.append(Paragraph("Xiaomi Note 15 5G", styles["CoverProduct"]))
story.append(Spacer(1, 10*cm))
story.append(PageBreak())

# CONTENT PARSER
bullet_buffer = []
prev_was_section = False

def flush_bullets():
    if not bullet_buffer:
        return
    items = [ListItem(Paragraph(md_to_rl(txt.strip()), styles["Bullet"])) for txt in bullet_buffer]
    story.append(ListFlowable(
        items,
        bulletType="bullet",
        bulletFontName="Helvetica",
        bulletFontSize=10.8,
        leftPadding=6,
        spaceBetween=3.5
    ))
    bullet_buffer.clear()

for line in lines:
    stripped = line.strip()

    # Skip separator / kosong
    if re.match(r"^[-—━•*_=]{3,}$", stripped) or not stripped:
        flush_bullets()
        story.append(Spacer(1, 5))
        continue

    # Section utama: "### "
    if stripped.startswith("### "):
        flush_bullets()
        if prev_was_section:
            story.append(PageBreak())
        else:
            story.append(Spacer(1, 8))

        section_title = stripped[4:].strip()
        story.append(Paragraph(md_to_rl(section_title), styles["Section"]))
        story.append(Spacer(1, 6))
        prev_was_section = True
        continue

    # Subsection: **...** atau ...:
    if (stripped.startswith("**") and stripped.endswith("**")) or \
       (stripped.endswith(":") and len(stripped) < 50 and not stripped.isupper()):
        flush_bullets()
        clean = stripped.strip("*").strip(":").strip()
        story.append(Paragraph(f"<b>{md_to_rl(clean)}</b>", styles["Subsection"]))
        story.append(Spacer(1, 5))
        continue

    # Bullet line
    if stripped.startswith(("-", "•", "*", "→", "–")) or re.match(r"^\s*[-*•→–]\s", line):
        bullet_text = re.sub(r"^[-*•→–]\s*", "", stripped).strip()
        bullet_buffer.append(bullet_text)
        continue

    # Paragraf biasa
    flush_bullets()
    story.append(Paragraph(md_to_rl(stripped), styles["Body"]))
    story.append(Spacer(1, 4))

flush_bullets()

# BUILD
doc.build(story)

print(f"PDF selesai → {OUTPUT_PDF.resolve()}")
print(f"Ukuran: {OUTPUT_PDF.stat().st_size / 1024:.1f} KB")
```
